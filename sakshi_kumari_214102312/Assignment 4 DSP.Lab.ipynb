# Question 7(a)

x(z)=z/(z+1/3)+z/(z-1/2)

```python
n=np.arange(-10,20,1)

def u_n(a,n):
    u_n1=[]
    for i in n:
        if i<a:
            u_n1.append(0)
        else:
            u_n1.append((-1/3)**i)
    return u_n1

def un(a1,n):
    u_n2=[]
    for i in n:
        if i<=a1:
            u_n2.append(-(1/2)**i)
        else:
            u_n2.append(0)
    return u_n2

un2=un(-1,n)


u_n2=u_n(0,n)
x_n=np.add(u_n2,un2) #function of x_n

plt.figure(figsize=(12,8))
plt.subplot(2,2,1)


plt.xlim([-10,20])

plt.stem(n,x_n)
plt.title('x[n]')

w=np.arange(0,20,1)

def x1_z():
    f= 1+np.exp(-1j*w)/3 # function for 1/1+e^(-jw/3)
    return 1/f
def x2_z():
    f1=1-np.exp(-1j*w)/2 # function for 1/1-e^(-jw/2)
    return 1/f1
y1=x1_z()
y2=x2_z()
x_z=np.add(y1,y2)

mag_x =np.absolute(x_z)
phase_x=np.angle(x_z)
plt.figure(figsize=(12,8))
plt.subplot(2,2,2)

plt.stem(w,mag_x)
plt.title('mag_x')

plt.figure(figsize=(12,8))
plt.subplot(2,3,3)

plt.stem(w,phase_x)
plt.title('phase_x')

```
