```python
n=np.arange(-30,50,1)              # range for time
xn=np.zeros(len(n))
N=20                               #given time period
def sig(N1):                       # function for x[n]
    for i in range(len(n)):
        if n[i]<=2+N1 and n[i]>=-2+N1:                   #N1 is the shift provided
            xn[i]=1       
        else:
            xn[i]=0
    return np.array(xn)
x_periodic=np.array(sig(0*N))+np.array(sig(N))+np.array(sig(2*N))+np.array(sig(-N))     #periodic signal by shifting method
plt.figure(figsize=(20,5))
plt.stem(n,x_periodic,'r')
plt.title('Periodic sequence x[n]')
plt.xlabel("Discrete time [n]")
plt.grid()
```

[1-b]

```python

n = np.arange(-10,11,1)
def DTFS(u,N):                          # here u is input signal for dtfs and N is period
    k = n.reshape((len(n), 1))          # creating an index values

    e = np.exp(-2j * np.pi * k * n / N) 
    #u=x_periodic
    def df(v):                            #summation function for DTFS formula
        sum=0
        for i in range(len(n)):
             sum=sum+(e[v][i]*u[i])
        return sum/N 
    dtfs=[] 
    for i in range(len(n)):
        dtfs.append(df(i))                        # collecting sum for each k value in a list i.e dtfs
    y=dtfs
    return y
y=DTFS(x_periodic,20)                       #calling dtfs function for signal,u=x_periodic and period=20 
fig, ax=plt.subplots(2,1,figsize=(15,11))        

ax[0].stem(n,np.abs(y),'g')                 #plotting magnitude and phase of coefficients  
ax[0].set_title('|Ck|')
ax[0].set_xlabel("Discrete time [n]")
ax[0].set_ylabel("Magnitude DTFS of x[n] ")
ax[0].grid()
ax[1].stem(n,np.angle(y),'b')
ax[1].set_title('/_Ck')
ax[1].set_ylabel("Phase DTFS of x[n] ")
ax[1].set_xlabel("Discrete time [n]")
ax[1].grid()
```

```python
N=20
def reconstruction(M):                     #defining common function for part a,b for reconstruction
    ck=[]                                  #collectinf fourier coeff values for M 
    y=DTFS(x_periodic,20)
    for i in range(-M,M+1):
        ck1=y[10+i]
        ck.append(ck1)
    n = np.arange(-10,10,1)
    n1 = n.reshape((len(n), 1))
    k=np.arange(-M,M+1,1)
    e = np.exp(2j * np.pi * k * n1 / N) 
    y=np.dot(e,ck)                        #performing Formula summation(e*ck)
    plt.figure(figsize=(15,6))
    plt.xlabel("Discrete time n")
    plt.title('Reconstructed x[n]')
    plt.grid()
    return  plt.stem(n,y,'purple')         #returning plot of reconstructed signal
```

```python
n=np.arange(-30,30,1)
l=np.pi 
w=np.linspace(-l,l,60)                #omega=[-pi,pi]          
w=w.reshape(len(w),1)              #indexing omega for calculations
xn=np.zeros(len(n))
e=np.exp(-1j*w*n)                #exp function

for i in range(len(n)):             #signal in figure(1)
        if n[i]== -2 or n[i]==0 or n[i]==1:
             xn[i]=1       
        elif n[i]==-1 or n[i]==2 :
             xn[i]=-1
        else:
             xn[i]=0
                
def DTFT(b):                   #DTFT function , where b=input signal
    def df(v):
        sum=0
        for i in range(len(n)):
            sum+=e[v][i]*b[i]         #summation  formula
        return sum
    dtft=[] 
    for i in range(len(n)):
        dtft.append(df(i))
    
    return dtft
plt.figure(figsize=(15,5))
plt.stem(n,xn,'r')
plt.grid()
plt.xlabel("Discrete time n")
plt.title('Fig(1):x[n]')
plt.show()

y=DTFT(xn)                           #y= DTFT of x[n], here b=x[n]
fig, ax=plt.subplots(1,2,figsize=(15,7))        
ax[0].plot(w,np.abs(y),color='orange')
ax[0].set_title('Magnitude spectra of X(w)')
ax[0].set_xlabel("Frequency (w)")
ax[0].grid()
ax[1].plot(w,np.angle(y),color='green')
ax[1].set_title('Phase spectra of X(w) ')
ax[1].set_xlabel("Frequency (w)")
ax[1].grid()
plt.show()
```

```python
n1=np.arange(-10,15,1)              
x=np.zeros(len(n1))

def input(a):                       #input signal function
    for i in range(len(n1)):
        
        if n1[i]<0:
            x[i]=0
        elif n1[i]>=(0):
            x[i]=(a)**n1[i]
    return np.array(x) 
xn=input(0.8)                         #x[n], for a=0.8
hn=input(1)                           #h[n], for a=1
n=np.arange(-10,39,1)
w=np.linspace(-l,l,len(n))
yn=np.convolve(xn,hn)            # performing convolution of x[n]*h[n]
dtft=DTFT(yn)                    #applying DTFT  on y[n] to get Y(w)
plt.figure(figsize=(15,7))
plt.plot(w,dtft,color='darkorange')
plt.title('DTFT of Y(w)')
plt.xlabel("Frequency w")
plt.grid()
plt.show()
```
