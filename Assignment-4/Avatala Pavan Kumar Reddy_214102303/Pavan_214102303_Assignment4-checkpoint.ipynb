```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal
import warnings
warnings.filterwarnings("ignore")
```

```python
N=20
def fun(n,T_p):
    k=[]
    for i in n:
        if np.abs(i-N*T_p)<=2: #x(n)=1 for -2<n<2
            k.append(1)
        else:
            k.append(0)
    return np.array(k)
n=np.arange(-10,10,1)
plt.stem(n,fun(n,0))
plt.title('x(n)')
plt.xlabel('n')
plt.show()
```

### a) Plot the sequence x[n] for 4 time periods.

```python
n=np.arange(-30,50,1)
x1=fun(n,0)
x2=fun(n,-1)
x3=fun(n,2)
x4=fun(n,1)
x=x1+x2+x3+x4

plt.figure(figsize=(12,8))
plt.stem(n,x)
plt.title('x(n) ,Time periods=4')
plt.xlabel('n')
plt.grid()
```

### (b) Determine and plot the magnitude, |ck| and phase coefficients, ∠ck of discrete time Fourier series (DTFS) of x[n].

```python
import cmath
n=np.arange(-N/2,N/2,1)
x=fun(n,0)
def c(k,x,N):
    n=np.arange(-N/2,N/2,1)
    x=np.array(x)
    l=np.exp(-1j*2*np.pi*k*n/N)
    return sum(x*l)/N
N=20
k=np.arange(-10,10,1)
y_mag=[np.abs(c(i,x,N)) for i in k]
plt.figure(figsize=(14,7))
plt.subplot(2,1,1)
plt.title('Magntiude |$c_{k}$|',size=15)
plt.ylabel('|$c_{k}$|',size=15)
plt.xlabel('k',size=15)
plt.stem(k,y_mag)

y_phase=[cmath.phase(c(i,x,N)) for i in k]
plt.subplot(2,1,2)
plt.title('Phase <$c_{k}$',size=15)
plt.ylabel('<$c_{k}$',size=15)
plt.xlabel('k',size=15)
plt.stem(k,y_phase)
plt.tight_layout()
plt.show()
```

### 2
#### (a) Reconstruct x[n] from 4 DTFS coefficients (i.e., M = 2) and plot the original signal (x[n]) and the reconstructed signal (ˆx[n]).
#### (b) Reconstruct x[n] from 10 DTFS coefficients (i.e., M = 5) and plot the original signal (x[n]) and the reconstructed signal (ˆx[n]).

```python
def recon_sig(M):
    N=20
    n=np.arange(-N/2,N/2,1)
    k=np.arange(-M,M+1,1)
    x=np.zeros(20)
    for j in k:
        C=c(j,fun(n,0),N) 
        x=x+C*np.exp(1j*2*np.pi*j*n/N)
        
    return np.array(x)
n=np.arange(-N/2,N/2,1) 
plt.figure(1,figsize=(14,7))
plt.subplot(211)
plt.stem(n,abs(recon_sig(2)))
plt.title('Reconstructed signal for M=2',size=15)
plt.xlabel('n',size=15)
plt.subplot(212)
plt.stem(n,abs(recon_sig(5)))
plt.title('Reconstructed signal for M=5',size=15)
plt.xlabel('n',size=15)
plt.tight_layout()
plt.show()
```

#### 4. Compute the Fourier transform X(ejω) of the signal x[n] shown in Fig.1 and plot its magnitude and phase over the interval ω ∈ [−π, π]


```python
import cmath
x=np.zeros(11) # -5 to 5
x[3]=1
x[4]=-1
x[5]=1
x[6]=1
x[7]=-1
def xdf(w,x):
    k=0
    n=np.arange(0,11,1)
    for i in n:
        k+=x[i]*np.exp(-1j*w*i)
    return k

plt.figure(1,figsize=(14,7))
w=np.linspace(-np.pi,np.pi,1000)
xdf_mag=[abs(xdf(i,x)) for i in w]
plt.subplot(2,1,1)
plt.plot(w,xdf_mag)
plt.title('Magnitude plot of X(w)',size=15)
plt.xlabel('rad/sample(w)',size=15)
plt.ylabel('|X(w)|',size=15)

xdf_phase=[cmath.phase(xdf(i,x)) for i in w]
plt.subplot(2,1,2)
plt.plot(w,xdf_phase)
plt.title('Phase plot of X(w)',size=15)
plt.xlabel('rad/sample(w)',size=15)
plt.ylabel('<X(w),radians',size=15)

plt.tight_layout()
plt.show()

```

### 5
####  Given an LTI system with impulse response h[n] = u[n] and an input x[n] = (0.8)nu[n], compute the DTFT of the output Y (e^jω) and its inverse DTFT y[n].


```python
u = lambda n: 1 if n>=0 else 0
def x(n):
    return np.power(0.8,n)*u(n)
h=u
plt.figure(figsize=(14,7))
n=np.arange(0,10,1)
plt.subplot(211)
h_n=[h(i) for i in n] #h(n)
plt.stem(n,h_n)
plt.title('h(n)',size=15)
plt.xlabel('n',size=15)
plt.subplot(212)
x_n=[x(i) for i in n] #x(n)
plt.stem(n,x_n)
plt.title('x(n)',size=15)
plt.xlabel('n',size=15)
plt.tight_layout()
plt.show()
```

```python
w=np.linspace(-np.pi,np.pi,10)
def DTFT(x,w,n):
    d=[]
    for i in w:
        X_w=np.dot(x,np.exp(-1j*i*n))
        d.append(X_w)
    return d
X_w=DTFT(x_n,w,n)
H_w=DTFT(h_n,w,n)

Y_w=[a*b for a,b in zip(X_w,H_w)]
print('The output Y(w) is:',Y_w)
```

```python
#defining funtion for the convolution
def convolution(inputsignal,impulseresponse):
        A=np.size(inputsignal) #length of input signal x[n]
        B=np.size(impulseresponse) #length of impulse response h[n]
        o = np.zeros(A+B-1)
        for a in range(A):
            for b in range(B):
                o[a+b] = o[a+b] + inputsignal[a]*impulseresponse[b]
        return o   
y_n=convolution(x_n,h_n)
plt.stem(y_n)
plt.title('y(n)')
plt.xlabel('n')
plt.show()
```

### 6
#### Use the DFT to compute approximately the first 64 Fourier coefficients of the function f(t) = (e^−t^2/10)[sin2t + 2 cos4t + 0.4 sintsin10t]  on the interval [0, 2π].


```python
from numpy import sin,cos,exp
x= lambda t:exp(-t**2/10)*(sin(2*t)+2*cos(4*t)+0.4*sin(t)*cos(10*t))
t=np.linspace(0,2*np.pi,64)
x=np.array([x(i) for i in t])
plt.stem(t,x)
print(x.shape)
```

```python
W=np.zeros((64,64),dtype='complex')
w=np.exp(-1j*(np.pi)/32)

for i in range(64):
    for j in range(64):
         W[i][j]=(w**(i*j))   
print(W.shape)
```

```python
y=np.dot(W,x)
plt.figure(figsize=(14,7))
plt.subplot(211)
plt.stem(np.abs(y))
plt.title('magnitude of coefficients',size=12)
plt.xlabel('frequency index k',size=12)
plt.ylabel('magnitude',size=15)
plt.subplot(212)
plt.stem(np.angle(y))
plt.title('phase of coefficients',size=12)
plt.xlabel('frequency index k',size=12)
plt.ylabel('phase in radians',size=15)
plt.tight_layout()
plt.show()
```
