```python
def fun(x):
    
    if -2<=x<=2:
        return 1
    else:
        return 0
```

```python
N1=10
N=20
x=np.arange(-N1,N1)
y=[]
for i in x:
    
    y.append(fun(i))
print(y)
plt.stem(x,y)
plt.title("x[n]")
plt.xlabel("n")
```

```python
k=np.arange(-10,10)
coeffs=[]

for i in k:
    s=0
    for j in x:
        s=s+fun(j)*np.exp(-i*2*np.pi*j*1J/N)
        
    coeffs.append(s/N)
    
# print(coeffs)
```

```python
mag=[]

for i in coeffs:
    mag.append(abs(i))
    
# print(mag)
```

```python
phase=[]

for i in coeffs:
    phase.append(np.arctan(i.imag/i.real))
    
# print(phase)
```

```python
ni=np.arange(-10,10)
plt.stem(ni,mag)
plt.title("|Ck|")
plt.xlabel("k")
```

```python
ni=np.arange(-10,10)
plt.stem(ni,phase)
plt.title("/_Ck")
plt.xlabel("k")
```

```python
def convolve(h,x):
   
    L=[]
    M=[]
    ans=[]
    for j in range(len(h)):
        for i in range(len(x)):
            L.append(h[j]*x[i])
        
        M.append(L)
        L=[]
#     print(M)
    
    def fun(ans,l,k):
        a=l
        b=k
        s=0
        while a<len(ans) and b>=0:
            s=s+ans[a][b]
            a=a+1
            b=b-1
        return s
    
    for i in range(len(M[0])):
        ans.append(fun(M,0,i))
    for j in range(1,len(M)):
        ans.append(fun(M,j,len(M[0])-1))
    
    return ans
```

```python
def fun(t):
    
    a=np.exp(-(t*t)/2)
    b=np.sin(2*t)+2*np.cos(4*t)+0.4*np.sin(t)*np.sin(10*t)
    
    return a*b
```

```python
t=np.linspace(0,2*np.pi,64)
f=fun(t)
plt.plot(t,f)
plt.title("plot of f(t)")
plt.xlabel("t")
plt.ylabel("amplitude")
```

```python
f=np.array(f).reshape(64,1)
```

```python
# print(f)
```

```python
import numpy as np
def DFT_matrix(N):
    i, j = np.meshgrid(np.arange(N), np.arange(N))
    omega = np.exp( - 2 * np.pi * 1J / N )
    W = np.power( omega, i * j ) 
    return W
```

```python
W=DFT_matrix(64)
```

```python
# W
```

```python
Y=np.matmul(W,f)
```

```python
k=np.arange(0,64,1)
```

```python
def PoleZeroPlot(b, a):
    (zeros,poles,gain) = tf2zpk(b, a)
    angle = np.linspace(-np.pi,np.pi,50)
    sin_x = np.sin(angle)
    cos_y = np.cos(angle)
    plt.figure()
    plt.plot(poles.real, poles.imag, 'x', zeros.real, zeros.imag, 'o', sin_x,cos_y, 'k-')
    circle1 = plt.Circle((0, 0), 0.5,color='r')
    plt.gca().add_patch(circle1)
    
    circle1 = plt.Circle((0, 0), 0.333,color='white')
    plt.gca().add_patch(circle1)
    
    plt.grid()
    plt.title("ploe zero plot")
    plt.xlim((-2, 2))
    plt.xlabel('Real axis')
    plt.ylim((-2, 2))
    plt.ylabel('Imaginary axis')
    plt.gcf().set_figwidth(5)
    return (zeros,poles,gain)
```
