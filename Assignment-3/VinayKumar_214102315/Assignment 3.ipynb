```python
#saw tooth wave function
def f(z):
    y = np.zeros(len(z))
    for i in range(len(z)):
        if z[i] < 0 and z[i] >=-1:
            y[i] += 1 + 2*z[i]
        else:
            y[i] +=  1 - 2*z[i]         
    return y
b = f(x)
```

```python
def fo(z):
    if z < 0 and z >=-1:
        y = (1 + 2*z)
    elif z >= 0 and z< 1:
        y = (1 - 2*z)         
    return y

Ao = quad(fo,-1,1)[0]/(2)

print("Value of DC part of fourier transform is {}".format(Ao))

def fc(z):
    if z < 0 and z >=-1:
        y = (1 + 2*z)*cos(i*z)
    elif z >= 0 and z< 1:
        y = (1 - 2*z)*cos(i*z)        
    return y

An= []

for i in range(50):
    an = quad(fc,-1,1)[0]
    An.append(an)

def fs(z):
    if z < 0 and z >=-1:
        y = (1 + 2*z)*sin(i*z)
    elif z >= 0 and z< 1:
        y = (1 - 2*z)*sin(i*z)      
    return y

Bn= []

for i in range(50):
    bn = quad(fs,-1,1)[0]
    Bn.append(bn)
    
fig, ax = plt.subplots(1,2, figsize = (20,10))
ax[0].stem(a,An,'b')
ax[0].set_xlabel("n")
ax[0].set_ylabel("An")
ax[0].set_title("Fourier series coeffcient An upto N= 50")
ax[1].stem(a,Bn,'r')
ax[1].set_xlabel("n")
ax[1].set_ylabel("An")
ax[1].set_title("Fourier series coeffcient Bn upto N= 50")
```
