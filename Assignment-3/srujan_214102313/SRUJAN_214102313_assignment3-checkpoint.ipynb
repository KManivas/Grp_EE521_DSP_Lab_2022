### Q.6. 
#### (a) Generate 3 sine waves with frequencies 2 Hz, 5 Hz, and 8 Hz, having amplitudes of 3, 2, and 1,  espectively, and all phases as zeros. Add the 3 sine waves together with a sampling rate fs = 100 Hz. Plot the resultant signal for a time duration of 1 sec.
#### (b) Write a function for discrete Fourier transform (DFT) as DFT(x) which takes in an input argument x, which is a 1-dimensional realvalued signal. The function has to calculate the DFT of the signal and return the DFT values. Apply this function to the signal generated in (a) and plot the result. Verify that the output of the DFT is symmetric.
#### (c) Filter out the DFT results to obtain only the frequency components of (a) and plot the resultant. That is, filter the frequency components which is greater than fs/2.

```python
def x1(t):
    return 3*np.sin(4*np.pi*t)
def x2(t):
    return 2*np.sin(10*np.pi*t)
def x3(t):
    return  np.sin(16*np.pi*t)
def x4(t):
    return x1(t)+x2(t)+x3(t)

t= np.arange(-10,10,0.01)
fig = plt.figure(1,figsize=(18,6))

plt.plot(t,x4(t))
def impulsetrain(n): 
    parabolic =[] 
    for sample in n:  
            parabolic.append(1) 
    return parabolic
n = np.arange(0,1,0.01)#100hz 
#x = impulsetrain(n)
fig = plt.figure(10,figsize=(60,20))
#plt.subplot(2,2,3)
#plt.stem(n,x,use_line_collection=True)  
#plt.title('Impulse train',color='red',size=20)
plt.subplot(2,2,4)
plt.stem(n,x4(n)*impulsetrain(n))
plt.tight_layout()
plt.show()
#print(n)
x=x4(n)*impulsetrain(n)
np.size(x)
```

```python
def DFT(x):

    N = len(x)
    n = np.arange(N)
    k = n.reshape((N, 1))
    e = np.exp(-2j * np.pi * k * n / N)
    
    X = np.dot(e, x)#summation over 0 to N-1(x(n)*e^(-j*2*pi*k*n/N))
    
    return X
```

```python
X = DFT(x)

# calculate the frequency
sr=100 #fs=100hz
N = len(X)
n = np.arange(N)
T = N/sr
freq = n/T 

plt.figure(figsize = (8, 6))
plt.stem(freq, np.abs(X), 'b', \
         markerfmt=" ", basefmt="-b")
plt.xlabel('Freq (Hz)')
plt.ylabel('DFT Amplitude |X(freq)|')
plt.grid()
plt.show()
```

```python
n_oneside = N//2 #Nyquist frequency
# get the one side frequency
f_oneside = freq[:n_oneside]

# normalize the amplitude
X_oneside =X[:n_oneside]/n_oneside #dividing by N/2

plt.figure(figsize = (12, 6))

plt.stem(f_oneside, np.abs(X_oneside), 'r', \
         markerfmt=" ", basefmt="-b")
plt.xlabel('Freq (Hz)')
plt.ylabel('DFT Amplitude |X(freq)|')


plt.xlim(0, 10)
plt.tight_layout()
plt.grid()
plt.show()
```
