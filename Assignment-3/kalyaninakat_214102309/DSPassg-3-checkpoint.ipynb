```python
import numpy as np
from numpy import sin,cos,pi,exp,sqrt
import matplotlib.pyplot as plt
from scipy.integrate import quad
from scipy.fftpack import fft,fftshift
```

[Question-1]

```python
T=2*pi

def A0(x):
    return 2*(pi-x)/T
I = quad(A0, 0,pi)              # using scipy library for integration
print("a0=",I[0])

def An(x):                        #an coeff function
    return 2*(pi-x)*cos(2*pi*x*i/T)/T
an=[]
for i in range(1,15):          # 15 coeffiecients im considering here
        Ia = quad(An, 0,pi)
        an.append(Ia[0])
def Bn(x):                       #bn coeff function
    return 2*(pi-x)*sin(2*pi*x*i/T)/T
bn=[]
for i in range(1,15):

        Ib = quad(Bn, 0,pi)
        bn.append(Ib[0])  
print("an=",an)    
print("bn=",bn)  
fig, ax=plt.subplots(1,2,figsize=(12,5))        
t=np.arange(1,15,1)
ax[0].stem(t,np.abs(an),'b')
ax[0].set_title("Magnitude of an")
ax[0].set_xlabel('n')
ax[0].set_ylabel('|a[n]|')
ax[0].grid()
ax[1].stem(t,np.abs(bn),'r')
ax[1].set_title("Magnitude of bn")
ax[1].set_xlabel('n')
ax[1].set_ylabel('|b[n]|')
ax[1].grid()
plt.show()
```

```python
T=2
def A0(x,m):                            #a0 coeff function
    return ((m*2*x)+1)/T
I = quad(A0, -1,0,args=(1))[0]+quad(A0, 0,1,args=(-1))[0]                 # using scipy for integration
a0=I


def An(x,m):                           #an coeff function
    return 2*((m*2*x)+1)*cos(2*pi*x*i/T)/T
an=[]
for i in range(1,25):
        Ia = quad(An, -1,0,args=(1))[0]+quad(An, 0,1,args=(-1))[0]
        an.append(Ia)
def Bn(x,m):                          #bn coeff function
    return 2*((m*2*x)+1)*sin(2*pi*x*i/T)/T
bn=[]
for i in range(1,25):

        Ib = quad(Bn, -1,0,args=(1))[0]+quad(Bn, 0,1,args=(-1))[0]
        bn.append(Ib)  

fig, ax=plt.subplots(1,2,figsize=(12,5))        #plotting an and bn coeff for given signal
t=np.arange(1,25,1)
ax[0].stem(t,np.abs(an),'b')
ax[0].set_title("Magnitude of an")
ax[0].set_xlabel('n')
ax[0].set_ylabel('|a[n]|')
ax[0].grid()
ax[1].stem(t,np.abs(bn),'r')
ax[1].set_title("Magnitude of bn")
ax[1].set_xlabel('n')
ax[1].set_ylabel('|b[n]|')
ax[1].grid()
plt.show()
```

```python
t=np.linspace(-1,1,100)
def partial(n):                        # partial sum for the 1st 'n' coeff
    sum=0
    for i in range(1,n+1):
         sum=sum+(an[i-1]*cos(pi*t*i))+(bn[i-1]*sin(pi*t*i))
    return sum+a0
plt.figure(figsize=(10,7))
plt.plot(t,partial(3),'r')           # calling partial function for n=3
plt.plot([-1,0,1],[-1,1,-1],'g')
plt.legend([r"Partial sum of order n=3",
            r"Original wave"
            ])
plt.xlabel('time(t)')
plt.ylabel('tiangular waveform and its patial sum')
plt.grid()
plt.show()
```

```python
A,a=2,4                                     # given values in question
t=np.arange(0,10,0.5) 
x=np.zeros(len(t)) 
def X():                                 # function for x(t)
    for i in range(len(t)):
        if t[i]<0:                                  
            x[i]=0
        elif t[i]>=0 :
            x[i]=A*(exp(-a*t[i]))
    return x
w=np.arange(-10,10,0.1)
def mag():                            # function for magnitude of x(w)    
    return 1/np.sqrt((w**2)+(a**2))
def angle():                          # function for phase of x(w) 
    return -np.arctan(w/a)
fig, ax=plt.subplots(1,3,figsize=(15,7))        
ax[0].plot(t,X(),'r') 
ax[0].set_title("x(t)")
ax[0].set_xlabel('time axis')
ax[0].grid()
ax[1].plot(w,mag(),'g') 
ax[1].set_title("|X(w)|")
ax[1].set_xlabel('freq')
ax[1].grid()
ax[2].plot(w,angle(),'b') 
ax[2].set_title("/_X(w)")
ax[2].set_xlabel('freq')
ax[2].grid()
```

[6-b]

```python
N = len(R)
n = np.arange(N)
k = n.reshape((N, 1))
def DFT(R):                  #function for DFT
    
    e = exp(-2j * pi * k * n / N)    
    X = np.dot(e, R) 
        
    return X
plt.figure(figsize=(10,7))
plt.stem(n,np.abs(DFT(R)),'r')          #calling DFT For resultant signal
plt.title("DFT of resultant signal")
plt.grid()
plt.show()
```

CONDITION FOR SYMMETRY:

The DFT of a real-valued discrete-time signal has a special symmetry, in which the real part of the transform values are DFT even symmetric and the imaginary part is DFT odd symmetric.

```python
fig, ax=plt.subplots(1,2,figsize=(10,6))        
ax[0].stem(n,np.abs(DFT(R)).real,'g') 
ax[0].set_title("real X(w)")
ax[0].grid()
ax[1].stem(n,DFT(R).imag,'g') 
ax[1].set_title("Img X(w)")
ax[1].grid()
```

From above plots we can say that give signal is symmetric

[7-c]

```python

x=np.zeros(len(u))
def window(tau):
    for i in range(len(u)):                  #function for window
        if u[i]<=(-tau) or np.abs(u[i])>=(tau):
            x[i]=0
        else:
            x[i]=1
    return np.array(x)
wind1=window(40)        #rectangular window with 80 samples
wind2=window(80)        #rectangular window with 160 samples
```

[7-d]

```python
fft1=fftshift(np.abs(fft(wind1)))             #using scipy for fft of window 1
fft2=fftshift(np.abs(fft(wind2)))            #using scipy for fft of window 2
fig, ax=plt.subplots(2,2,figsize=(20,15))        
ax[0][0].stem(u,wind1,'g') 
ax[0][0].set_title("80 samples rect window")
ax[0][0].grid()
ax[0][1].plot(t,fft1,'g') 
ax[0][1].set_title("Frequency domain for 80 sample rect window")       #plotting its fft
ax[0][1].grid()
ax[1][0].stem(u,wind1,'r') 
ax[1][0].set_title(" 160 sample rect window")
ax[1][0].grid()
ax[1][1].plot(t,fft2,'r') 
ax[1][1].set_title(" Frequency domain for 160 sample rect window")
ax[1][1].grid()
```

For 80 samples main and side lobes of sinc function is distorted.
For 160 samples main and side lobes of sinc function is are smooth.
