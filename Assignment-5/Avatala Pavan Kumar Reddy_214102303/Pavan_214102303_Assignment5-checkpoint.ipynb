## 1
#### Generate the following signal for 1 sec duration at sampling frequency of 200 Hz.
#### x(t) = 9 sin (2πf1t) + 7 sin (2πf2t) + 5 sin (2πf3t) + 3 sin (2πf4t) where f1 = 5 Hz, f2 = 15 Hz, f3 = 25 Hz and f4 = 50 Hz.

```python
# sampling rate
fs=200
# sampling interval
ts=1.0/fs
t=np.arange(0,1,ts)

f1=5
f2=15
f3=25
f4=50
x=9*sin(2*pi*f1*t)+7*sin(2*pi*f2*t)+5*sin(2*pi*f3*t)+3*sin(2*pi*f4*t)

plt.figure(figsize = (14, 7))
plt.stem(t, x)
plt.ylabel('Amplitude')
plt.xlabel('Time')


plt.show()
```

#### b).Determine the DFT of x(t) and plot its 1-sided magnitude spectrum (with frequency in Hz on x-axis

```python
def DFT(x):
    N=len(x)
    n=np.arange(N)
    k=n.reshape((N, 1))
    w=np.exp(-2j*np.pi*k*n/N)
    X=np.dot(w,x)    
    return X

X=DFT(x)

# calculate the frequency
N=len(X)
n=np.arange(N)
T=N/fs
freq=n/T 

n_oneside = round(N/2)
# get the one side frequency
f_oneside = freq[:n_oneside]
X_oneside =X[:n_oneside]

plt.figure(figsize = (12, 6))
plt.stem(f_oneside, abs(X_oneside))
plt.title('Onesided DFT')
plt.xlabel('Hz')
plt.ylabel('DFT Amplitude |X(freq)|')

plt.show()
```

#### (c) Generate a digital butterworth bandpass filter of order-6 (using scipy) with 10 Hz and 30 Hz as its lower and upper cutoff frequencies, respectively, and plot its magnitude spectrum.

```python
from scipy.signal import butter
from scipy.signal import freqz
lower_cutoff=10
upper_cutoff=30
nyquest=0.5*fs
lower=lower_cutoff/nyquest
upper=upper_cutoff/nyquest
b, a = butter(6, [lower, upper], btype='band')
w, h = freqz(b, a, worN=100)
plt.plot(abs(h))
plt.title('Bandpass Filter')
plt.xlabel('Hz')
plt.ylabel('Magnitude')
plt.show()
```

#### (d) Multiply the frequency response of bandpass filter and DFT spectrum of x(t) to obtain the filtered spectrum and plot the same.


```python
filtered=h*X_oneside
plt.stem(abs(filtered))
plt.title('filtered spectrum')
plt.xlabel('Hz')
plt.show()
```

#### (e) Apply inverse DFT on the filtered spectrum to obtain the filtered signal and plot the same.


```python
def invDFT(x):
    N=len(x)
    n=np.arange(N)
    k=n.reshape((N, 1))
    w=np.exp(2j*np.pi*k*n/N)
    X=np.dot(w,x)/N
    return X
t=np.linspace(0,1,100)
plt.plot(t,invDFT(filtered))
plt.title('Filtered wave')
plt.xlabel('time')
plt.ylabel('Amplitude')
plt.grid()
plt.show()
```

#### Consider a random sequence x[n] of length N = 1024.
#### (a) Write a function to compute the DFT of the sequence. How many operations are required for computing the DFT? Don’t use any inbuilt function for this part. After implementation, compute the time taken by the code to run using the time library.
#### (b) Compute the DFT of sequence using np.fft. Compute the time taken by the code to run using the time library.
#### (c) Are the implementations in part (a) and part (b) similar? Compare the closeness and compute time for both the parts? What can be concluded?
#### (d) Vary the number of samples N from 1, 2, 4, 8, ..., 2048. For each N compare the performance in terms of time for DFT implemented in part (a) and FFT implemented in part (b). Plot a graph for the same

```python
def xn(n):
    k=[]
    for i in range(n):
        k.append(np.random.random())
    return k
def DFT(x):
    N=len(x)
    n=np.arange(N)
    k=n.reshape((N, 1))
    w=np.exp(-2j*np.pi*k*n/N)
    X=np.dot(w,x)    
    return X
x=xn(1024)
from time import perf_counter
def time_DFT(xn):
    start = perf_counter()
    DFT(xn)
    end = perf_counter()
    return end-start

time=time_DFT(x)
print("The time of execution of above program is :",{time})
```

```python
from time import perf_counter
def time_FFT(xn):
    start = perf_counter()
    np.fft.fft(xn)
    end = perf_counter()
    return end-start

time=time_FFT(x)
print("The time of execution of above program is :",{time})
```

```python
import time

dft=[]
fft=[]
k=1
while k<=2048:
    x=xn(k)
    dft.append(time_DFT(x))
    fft.append(time_FFT(x))
    k=k*2
N=[2**i for i in range(12)]
plt.plot(N,dft,label='time of DFT')
plt.plot(N,fft,label='time of FFT')
plt.xlabel('Input size')
plt.ylabel('time')
plt.legend()
plt.show()
```

### Decimation in Time (DIT)
#### Consider a sequence x[n] = {1, 1, 1, 1, 0, 0, 0, 0}, compute the DFT of the following sequence using 8-point radix-2 DIT-FFT algorithm. Plot the sequence and the resulting DFT. Also compute the time taken to run the code.

```python
import cmath
import matplotlib.pyplot as plt
exp = np.exp
from time import perf_counter


#input
x0 = [1,1,1,1,0,0,0,0]
#initialize the outputs of three stages 
x = [0,0,0,0,0,0,0,0]

x1 = x.copy()
x2 = x.copy()
x3 = x.copy()
#  Finding out the twiddle Factor

w=exp(-1j*2*pi/8)
w0=1
w1=w
w2=w*w
w3=np.power(w,3)
start=perf_counter()
#  Stage 1

n=len(x1)
for i in range(n):
    if(i<n/2):
        if(i%2==0):
            x1[i]=x0[i]+x0[int((n/2+i)%8)]
        else:
            x1[i]=x0[i-1]-x0[int((n/2+i)%8)]
    else:
        if(i%2==0):
            x1[i]=x0[int((n/2+i+1))%8]+x0[i+1]
        else:
            x1[i]=x0[int((n/2+i)%8)]-x0[i]


#  Stage 2

x2[0] = x1[0] + w0 * x1[2]
x2[1] = x1[1] + w2 * x1[3]
x2[2] = x1[0] - w0 * x1[2]
x2[3] = x1[1] - w2 * x1[3]

x2[4] = x1[4] + w0 * x1[6]
x2[5] = x1[5] + w2 * x1[7]
x2[6] = x1[4] - w0 * x1[6]
x2[7] = x1[5] - w2 * x1[7]
#  Stage 3
for i in range(n):
    if(i<n/2):
        x3[i]=x2[i]+(w**i)*x2[int(n/2+i)]
    else:
        x3[i]=x2[int(i-n/2)]-(w**(i-n/2))*x2[i]



end=perf_counter()
X=[round(ele,2) for ele in x3]
print('Output:',X)
print('Time taken os:',end-start)
n=np.arange(8)
plt.subplot(2, 1, 1)
#plt.plot(n, [0, 0, 0, 0, 0, 0, 0, 0], color='black')  # Just for Reference
plt.plot(n, x0, 'bo')
plt.tight_layout()
plt.xlabel('time[n]')
plt.ylabel('Value x[n]')
plt.grid()

plt.subplot(2, 1, 2)
#plt.plot(disc, [0, 0, 0, 0, 0, 0, 0, 0], color='black')  # Just for Reference
plt.plot(n, abs(np.array(x3)), 'ro')
plt.tight_layout()
plt.xlabel('cycles/sample')
plt.ylabel('Value X(n)')
plt.grid()
plt.show()
```

```python
fig = plt.figure(1,figsize=(10,7))
#defining funtin for the convolution
def convolution(inputseq,impulseresponse):
        lengthA=np.size(inputseq)
        lengthB=np.size(impulseresponse)
        C = np.zeros(lengthA + lengthB -1)
        for m in np.arange(lengthA):
            for n in np.arange(lengthB):
                C[m+n] = C[m+n] + inputseq[m]*impulseresponse[n]
        return C 
x=[1,2,3,1] #input sequences
h=[1,1,1]
n = np.arange(-10,10,1)
p=convolution(x,h) #performing convolution
print("output sequence=",p) #printing the convolution result
fig = plt.figure(1,figsize=(10,7))
plt.subplot(3,1,1)
plt.stem(p,use_line_collection=True) #plotting convolution result.
plt.title('output')
plt.xlabel('n')
plt.subplot(3,1,2)
plt.stem(x,use_line_collection=True)
plt.title('x(n)')
plt.xlabel('n')
plt.subplot(3,1,3)
plt.stem(h,use_line_collection=True)
plt.title('h(n)')
plt.xlabel('n')
plt.tight_layout()
plt.show()
```

```python
x=[1,2,3,1]
h=[1,1,1]
m=len(x)
n=len(h)
def shifter(col):
    last = col[len(col)-1]
    x = len(col)
    result = [0] * x
    for i in range(1,len(col)):
        result[i] = col[i-1]
        result[0] = last
    return result
#  finding circular convolution
def circularconv(x,h,n,m):
    mat=np.zeros((max(n,m),max(n,m)))
    for i in range(0,len(mat[0])):
        mat[0][i]=x[i]
    for i in range(1,max(n,m)):
        mat[i]=shifter(mat[i-1])
        matrix=np.transpose(mat)
        diff = abs(n-m)
    for i in range(m,m+diff):
        h.append(0)
        resultant = np.dot(matrix,h)
    return resultant
circular=circularconv(x,h,m,n)
print("output sequence=",circular) #printing the convolution result
plt.figure(figsize=(10,8))
plt.subplot(3,1,1)
plt.stem(circular,use_line_collection=True) #plotting convolution result.
plt.title('output')
plt.xlabel('n')
plt.subplot(3,1,2)
plt.stem(x,use_line_collection=True)
plt.title('x(n)')
plt.xlabel('n')
plt.subplot(3,1,3)
plt.stem(h,use_line_collection=True)
plt.title('h(n)')
plt.xlabel('n')
plt.tight_layout()
plt.show()
```
