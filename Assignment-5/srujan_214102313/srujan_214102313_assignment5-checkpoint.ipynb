```python
def x1(t,f1,f2,f3,f4):
    x1=[]
    for i in t:
        x1.append((9*sin(2*np.pi*f1*i))+(7*sin(2*np.pi*f2*i))+(5*sin(2*np.pi*f3*i))+(3*sin(2*np.pi*f4*i)))
    return x1
fs=200
ts=1/fs
t=np.arange(0,1,ts)
fig = plt.figure(1,figsize=(18,6))
plt.stem(t,x1(t,5,15,25,50))
plt.title('x1(t)')
plt.show()
```

```python
def DFT(x):

    N = len(x)
    n = np.arange(N)
    k = n.reshape((N, 1))
    e = np.exp(-2j * np.pi * k * n / N)
    
    X = np.dot(e, x)#summation over 0 to N-1(x(n)*e^(-j*2*pi*k*n/N))
    
    return X
```

```python
X = DFT(x1(t,5,15,25,50))

# calculate the frequency
sr=200 #fs=200hz
N = len(X)
n = np.arange(N)
T = N/sr
freq = n/T 

plt.figure(figsize = (8, 6))
plt.stem(freq, np.abs(X), 'b', \
         markerfmt=" ", basefmt="-b")
plt.xlabel('Freq (Hz)')
plt.ylabel('DFT Amplitude |X(freq)|')
plt.grid()
plt.show()
```

```python
n_oneside = N//2 #Nyquist frequency
# get the one side frequency
f_oneside = freq[:n_oneside]

# normalize the amplitude
X_oneside =X[:n_oneside]/n_oneside #dividing by N/2

plt.figure(figsize = (12, 6))

plt.stem(f_oneside, np.abs(X_oneside), 'r', \
         markerfmt=" ", basefmt="-b")
plt.xlabel('Freq (Hz)')
plt.ylabel('DFT Amplitude |X(freq)|')


plt.xlim(0, 100)
plt.tight_layout()
plt.title('One sided spectrum')
plt.grid()
plt.show()
```

**butterworth bandpass filter**

```python
def butter_bandpass(lowcut, highcut, fs, order=6):
    nyq = 0.5 * fs
    low = lowcut / nyq
    high = highcut / nyq
    b, a = butter(order, [low, high], btype='band')
    return b, a
```

```python
b, a = butter_bandpass(10, 30, 200, order=6)
w, h = freqz(b, a, worN=100)
plt.plot((fs * 0.5 / np.pi) * w, abs(h), label="order = 6")
plt.xlabel('f')
plt.title('Bandpass')
plt.show()
```

```python
filter_spec=h*X_oneside
plt.stem(f_oneside,abs(filter_spec))
plt.xlabel('f')
plt.title('filtered spectrum')
plt.show()
```

```python
def IDFT(x):

    N = len(x)
    n = np.arange(N)
    k = n.reshape((N, 1))
    e = np.exp(2j * np.pi * k * n / N)
    
    X = np.dot(e, x)#summation over 0 to N-1(x(n)*e^(j*2*pi*k*n/N))
    
    return X
 
inverse=IDFT(filter_spec)
```

```python
plt.plot((inverse))
plt.xlabel('t')
plt.title('filtered signal')
plt.show()
```

**6)FFT**

```python
def x_n(n):
    x_n=[]
    for i in range(n):
        x_n.append(random.random())
    return x_n
```

```python
def DFT(x,N):

    #N = len(x)
    n = np.arange(N)
    k = n.reshape((N, 1))
    e = np.exp(-2j * np.pi * k * n / N)
    
    X = np.dot(e, x)#summation over 0 to N-1(x(n)*e^(-j*2*pi*k*n/N))
    
    return X
x_1024=x_n(1024)
```

```python
from time import perf_counter
def DFT_tym(n):
    start = perf_counter()
    DFT(x_n(n),n)
    end = perf_counter()
    return end-start

# difference of start and end variables
# gives the time of execution of the
# program in between
time=DFT_tym(1024)
print("The time of execution of above program is :",{time})
```

```python
from time import perf_counter
def FFT_tym(n):
    start = perf_counter()
    np.fft.fft(x_n(n))
    end = perf_counter()
    return end-start
# difference of start and end variables
# gives the time of execution of the
# program in between
time=FFT_tym(1024)
print("The time of execution of above program is :",{time})
```

**a)linear convolution**

```python
fig = plt.figure(1,figsize=(10,7))
#defining funtin for the convolution
def convolution(inputseq,impulseresponse):
        lengthA=np.size(inputseq)
        lengthB=np.size(impulseresponse)
        C = np.zeros(lengthA + lengthB -1)
        for m in np.arange(lengthA):
            for n in np.arange(lengthB):
                C[m+n] = C[m+n] + inputseq[m]*impulseresponse[n]
        return C 
a1=[1,2,3,1] #input sequences
b1=[1,1,1]
n = np.arange(-10,10,1)
p=convolution(a1,b1) #performing convolution
print("output sequence=",p) #printing the convolution result
plt.subplot(3,1,1)
plt.stem(p,use_line_collection=True) #plotting convolution result.
plt.title('output',color='red',size=20)
plt.subplot(3,1,2)
plt.stem(a1,use_line_collection=True)
plt.title('input',color='red',size=20)
plt.subplot(3,1,3)
plt.stem(b1,use_line_collection=True)
plt.title('impulse sequence',color='red',size=20)
plt.tight_layout()
plt.show()
```

**b)Circular convolution**

```python
x=[1,2,3,1]
h=[1,1,1]
m=len(x)
n=len(h)
```

```python
#to get the shifts in each column
def shifter(matrix):
    last = matrix[len(matrix)-1]
    x = len(matrix)
    result = [0] * x
    for i in range(1,len(matrix)):
        result[i] = matrix[i-1]
        result[0] = last
    return result
```

```python
#  finding circular convolution
def circularconv(x,h,n,m):
    ini_matrix = np.zeros((max(n,m),max(n,m)))
    for i in range(0,len(ini_matrix[0])):
        ini_matrix[0][i] = x[i]
    for i in range(1,max(n,m)):
        ini_matrix[i] = shifter(ini_matrix[i-1])
    final_matrix = np.transpose(ini_matrix)
    diff = abs(n-m)
    for i in range(m,m+diff):#for zero padding
        h.append(0)
    resultant = np.dot(final_matrix,h)
    return final_matrix,resultant
```

```python
final_matrix,circular=circularconv(x,h,m,n)
```
