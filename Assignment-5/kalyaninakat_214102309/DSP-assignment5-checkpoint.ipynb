```python
t=np.linspace(0,1,100)
x_t=9*sin(10*pi*t)+7*sin(30*pi*t)+5*sin(50*pi*t)+3*sin(100*pi*t)  #given signal
```

[1-a]

```python
plt.figure(figsize=(15,7))   #plotting given signal
plt.plot(t,x_t,'r')
plt.title('X(t) signal')
plt.xlabel('time t (in sec)')
plt.grid()
plt.show()
```

[1-b]


```python
N=len(x_t)
n = np.arange(N)
k = n.reshape((N, 1))
def DFT(x):                  #function for DFT
    
    e = exp(-2j * pi * k * n / N)    
    X = np.dot(e, x) 
        
    return X
print(DFT(x_t))    #printing dft of x(t)
```

```python
w=np.linspace(0,100,N)      #freq in hertz
plt.figure(figsize=(15,7))   #plotting magnitude of dft
plt.stem(w,np.abs(DFT(x_t)),'maroon')
plt.title('|X(w)|')
plt.xlabel('freq in Hz')
plt.grid()
plt.show()
```

```python
time_start = time.perf_counter()       #start time calculation using time library
x_n = [1,1,1,1,0,0,0,0]
#print("List Items = ", evenList)
N=8
x_e=[]
x_o=[]
for i in range(N):
    if (i % 2 == 0):
        x_e.append(x_n[i])
    else:
        x_o.append(x_n[i])
W=exp(-2j*np.pi/8)
G=np.fft.fft(x_e)                   #DFT of even elements
H=np.fft.fft(x_o)                    #DFT of odd elements
out=[]
for i in range(N):
    if i<=3:
        y=G[i]+((W**i)*H[i])
        out.append(y)
    else:
        y=H[i-4]+((W**i)*G[i-4])
        out.append(y)
        
fig, ax=plt.subplots(1,2,figsize=(15,7))        
ax[0].stem(np.arange(0,8,1),x_n,'purple')
ax[0].set_title('x[n]')
ax[0].set_xlabel("Discrete time (n)")
ax[0].grid()
ax[1].stem(np.arange(0,8,1),np.abs(out),'yellow')
ax[1].set_title('DFT of x[n] using radix-2 DIT-FFT')
ax[1].set_xlabel("K")
ax[1].grid()
plt.show()
time_end = time.perf_counter()         #end time calculation using time library
run_time = time_end - time_start
print("Time taken to run the code:",run_time,"sec")
```

[9-a]:LINEAR CONVOLUTION

```python
l = [1,2,3,1]
h=[1,1,1,0]
L=len(l)+len(h)-1
def linear_con(p,q):                                                       #CONVOLUTION FUNCTION DEFINED HERE  
    q_new=np.pad(q,(0,L-len(q)))
    sum=np.zeros(L)
    for i in range(len(p)):
        sum=sum+p[i]*q_new
        q_new=np.roll(q_new,1)
    return sum  
print("Linear convolution:",linear_con(l,h))
```

[9-b]:CIRCULAR CONVOLUTION

```python
def rotate(l, n):                     #performing matrix approach for circular convolution
     return l[n:] + l[:n]
N=len(l)
W=np.zeros((N,N))
for i in range(N):
    W[:,i]=rotate(l,N-i) 
CC= np.dot(W,h)
print("Circular convolution:",CC)
```

Property of DFT to implement circular convolution:

Circular convolution of x(n) & h(n) is equal to multiplication of their DFT s.
