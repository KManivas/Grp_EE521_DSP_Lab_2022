```python
def fact(num):                                            #defining function for factorial
    factorial =1
    if num ==0:
        factorial=1
    else:
        for i in range(1, num+1):
            factorial = factorial * i
    return factorial
def comb(n,k):                                          #defining function for combination formula used in probability
    C=fact(n)/(fact(n-k)*fact(k))
    return C
def binomial(n,k,p):                                   #defining function for formula of binomial dist
    pr=comb(n,k)*(p**k)*((1-p)**(n-k))
    return pr
with_ecc=binomial(7,0,0.1)+binomial(7,1,0.1)          # for x=0 or 1 ; here ecc:error correcting code
without_ecc=binomial(4,0,0.1)                         #for x=0
A=['with_ecc','without_ecc']
B=[with_ecc,without_ecc]
plt.figure(figsize=(9,5))
plt.bar(A, B,color='orange')
plt.ylabel('Probability')
plt.show()
```

```python
n=np.arange(-15,15,1)
x=np.zeros(len(n))
N=8
fig, ax=plt.subplots(1,2,figsize=(10,6))                 # for subplot and its fig size      

def imp(a,b,c):                                          #for x(n)
    for i in range(0,len(n)):
            if (n[i])>=0+c and (n[i])<=(N-1+c):         # c:shift  
                x[i]=a
            else:
                x[i]=0
               
    return x
h1=imp(1,1,0)                                           
def auto(d):                                         # function for autocorreletion
    sum=0
    for i in range(8):
        sum=sum+(d*imp(1,1,-i))                       #summation of x(n)*x(n-k); from 0 to N-1
        rk=sum/N
    return rk
ax[0].stem(n,h1,'r')
ax[0].set_title("Input sample x(n)")
ax[0].grid()
ax[1].stem(n,auto(h1),'r')
ax[1].set_title("Autocorreletion sample of x(n)")
ax[1].grid()
plt.show()
```

```python
def AC(x):
    N = len(x)
    r = np.zeros(N)
    for k in range(N):
        for i in range(N):
            if(i-k<0):
                r[k] += 0
            else:
                r[k] += (x[i]*x[i-k])/N
    return r
N = 1000
x = np.random.randn(N)
M = 100
r = AC(x)[0:M]           # First M lags of autocorrelation
true_autocorr = [1] + [0 for i in range(M-1)]
n = [i for i in range(M)]

fig,ax= plt.subplots(1,2,figsize=(10,6))
ax[0].stem(n,r,'g')
ax[0].set_title("First M lags autocorreltion")
ax[0].grid()

ax[1].stem(n,true_autocorr)
ax[1].set_title("True autocorrelation")
ax[1].grid()
```

```python
def average(x,M):
    N = len(x)
    r = np.zeros(M)
    for k in range(M):
        for m in range(10): # 10 different samples
            for n in range(M):
                if(n-k+100*m<0):
                    r[k] += 0
                else:
                    r[k] += (x[n+100*m]*x[n-k+100*m])/N
    return r
    
x = np.random.randn(N)
r_avg = average(x,100)
plt.stem(n,r_avg,'r')
plt.grid()
plt.show()
```

```python
x=np.linspace(0,20,100)
import math
df=[1,2,3,4,5]
def chi(k):                                                             #chi square pdf function
    pdf=(x**((0.5*k)-1))*(np.exp(-x/2))/((2**(k/2))*math.gamma(k/2))    # its formula
    return pdf
plt.figure(figsize=(10,7))
plt.plot(x,chi(1),color='red')                                       #degree of freedom=[1,2,3,4,5]
plt.plot(x,chi(2),color='blue')
plt.plot(x,chi(3),color='green')
plt.plot(x,chi(4),color='purple')
plt.plot(x,chi(5),color='orange')
plt.legend([r"df=1",
            r"df=2",
            r"df=3",
            r"df=4",
            r"df=5"])
plt.title("Chi square distribution")
plt.grid()
plt.show()
```

```python
x=np.linspace(0,20,100)
def I(k,l):                                        #bessel function used in pdf of non-central chi sq dist
    v=(k/2)-1
    y=np.sqrt(l*x)
    sum=0
    for i in range(100):
        sum=sum+(((y**2)/4)**i/(fact(i)*math.gamma(v+i+1)))
    C=((y/2)**v)*sum
    return C
def pdf(k,l):                                                    #non-central chi sq dist pdf
    pdf=0.5*(np.exp(-(x+l)/2))*((x/l)**(0.25*k-0.5))*I(k,l)      # its formula
    return pdf

plt.figure(figsize=(10,7))
plt.plot(x,pdf(2,1),color='red')                                # for k=2,l=[1,2,3]
plt.plot(x,pdf(2,2),color='blue')
plt.plot(x,pdf(2,3),color='green')
plt.plot(x,pdf(4,1),color='purple')                             # for k=4,l=[1,2,3]
plt.plot(x,pdf(4,2),color='orange')
plt.plot(x,pdf(4,3),color='black')
plt.legend([r"K=2 and lambda=1",
            r"K=2 and lambda=2",
            r"K=2 and lambda=3",
            r"K=4 and lambda=1",
            r"K=4 and lambda=2",
           r"K=4 and lambda=3"])
plt.title("Non-central Chi square distribution")
plt.grid()
plt.show()
```
