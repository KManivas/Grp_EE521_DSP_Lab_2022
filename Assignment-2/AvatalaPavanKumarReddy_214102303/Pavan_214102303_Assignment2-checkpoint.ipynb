##### Plot the probability of using error-correcting codes and the probability of without using error-correcting codes.

## 3.Autocorrelation

$$ With \ x(n) = 1 \ for\ n = 0, 1, . . . , 7; \ plot\ the\ input\ sequence\ and\ the\
sample\ autocorrelation \ sequence \  r_x $$

$$ r_x(k)\ = \frac 1N \sum_{n=0}^{N-1}\ x(n)\ x^*(n-k) $$

```python
x=np.array([1,1,1,1,1,1,1,1]) #given in the question
N=len(x)
def autocorrelation(x):
    N=len(x)
    r=[]          
    k=x.copy()
    for i in range(N):
        # code for positive half of autocorrelation function
        if i==0:
            r.append(sum(x*k)/N)
        else:
            k=np.insert(k,i-1,0) #if i=1 r[0] = x(n)x(n-1) 0 is added at the start to get the shift n-1
            x=np.insert(x,N,0) #0 is added at the last of x(n) to avoid index error caused by shift n-1
            r.append(sum(x*k)/N) # r gives positive half of the autocorrelation function
            
    #code for bulding auto correlation function from positvi half        
    r=np.concatenate([np.flip(r[1::]),r]) #revesing r and adding it to r gives the full autocorrelation since it is odd function
    return r
r=autocorrelation(x)

fig=plt.figure(1,figsize=(14,7))
plt.subplot(2,1,1)
plt.stem(x,use_line_collection=True)
plt.title('Input sequence x(n)',size=12)
n=np.arange(-N+1,N,1)
plt.subplot(2,1,2)
plt.stem(n,r)
plt.title('sample autocorrelation r(n)',size=12)

plt.show()
```

#### Correlation coefficient is given by:
$$ r_{xy} = \frac {\sum_{i=1}^{n} (x_{i}-\hat x)(y_{i}-\hat y)}{\sqrt {\sum_{i=1}^{n} (x_{i}-\hat x)^2}\sqrt{{\sum_{i=1}^{n}(y_{i}-\hat y)^2}}} $$
#### Covariance is given by:
$$ Cov_{xy} = \frac {\sum_{i=1}^{n} (x_{i}-\hat x)(y_{i}-\hat y)}{N-1} $$

```python
def correlation_coeff(x,y):
    x_m=np.mean(x)
    x_den=np.sqrt(np.sum((x-x_m)**2))
    y_m=np.mean(y)
    y_den=np.sqrt(np.sum((y-y_m)**2))
    den=x_den*y_den
    num=np.sum((x-x_m)*(y-y_m))
    return num/den
                  
def covariance(x,y):
    N=len(x)
    x_m=np.mean(x)
    y_m=np.mean(y)
    den=N-1
    num=np.sum((x-x_m)*(y-y_m))
    return num/den
```

##### The pdf of a central chi squared distribution is given by
$$ f(x) = \frac{1}{2^{n/2} \Gamma(n/2)} x^{n/2-1} e^{-x/2}$$

##### The pdf of a non-central chi squared distribution is given by
$$ g(x) = \sum_{k=0}^{\infty} e^{-\lambda/2} \frac{(\lambda/2)^{k}}{k!} f_{n+2k}(x)$$  
##### where $ f_{k}(x) $ represent pdf of chi squared distribution with k degrees of freedom
