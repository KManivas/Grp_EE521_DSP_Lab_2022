### Basic Signals

```python
t=np.linspace(-10,20,100)
N=len(t)
def u(t):
    N1=len(t)
    x=np.zeros(N1)
    for i in range(N1):
        if t[i]>=0:
            x[i]=1 
    return x
plt.plot(t,u(t))
plt.title('Unit Step Signal')
```

```python
def ramp(t):
    return t*u(t)
    
#x2=t*x1
plt.plot(t,ramp(t))
plt.title('Ramp Signal')
```

```python
def para(t):
    return t*t*u(t)/2
#x3=(t**2)*x1/2
plt.plot(t,para(t))
plt.title('Parabolic Signal')
```

```python
def rect(t,tow):
    x=np.zeros(len(t))
    for i in range(0,len(t)):
        if -tow/2<t[i]<tow/2:
            x[i]=1
    return x
plt.plot(t,rect(t,10))
plt.title('Rectangular Signal')
```

```python
def x5(t):
    return conv(rect(t,10),rect(t,10))
t5=np.linspace(-20,40,199)
plt.plot(t5,x5(t))
```

```python
x6=np.sinc(t)
plt.plot(t,x6)
```

### LTI Systems

```python
def conv(xn1,hn1):
    reslen=len(xn1)+len(hn1)-1
    xn2=np.pad(xn1,(0,reslen-len(xn1)))
    hn2=np.pad(hn1,(0,reslen-len(hn1)))
    resn=np.zeros(reslen)
    for i in range(reslen):
        resn2=[]
        for j in range(i+1):
            resn2.append(xn2[j]*hn2[i-j])
        #print(resn2)
        resn[i]=np.sum(resn2)
    return resn
```

```python
xn1=[3,5,2,6,8]
hn1=[-2,4,1]
conv(xn1,hn1)
```

```python
tn=np.arange(0,reslen,1)
plt.stem(tn,xn2,'b',markerfmt='bo',label='x[n]')
plt.stem(tn,hn2,'r',markerfmt='ro',label='h[n]')
plt.stem(tn,conv(xn1,hn1),'g',markerfmt='go',label='y[n]=x[n]*h[n]')
plt.legend()
```

```python
xn3=[2,3,4,7,1]
hn3=[-2,4,1]
hni=hn3[::-1]
conv(xn3,hni)
```

```python
tn2=np.arange(-2,5,1)
hn4=np.pad(hni,(0,len(tn2)-len(hni)))
xn4=np.pad(xn3,(0,len(tn2)-len(xn3)))
plt.stem(tn,xn4,'b',markerfmt='bo',label='x[n]')
plt.stem(tn2,hn4,'r',markerfmt='ro',label='h[n]')
plt.stem(tn2,conv(xn3,hni),'g',markerfmt='go',label='y[n]=x[n]*h[n]')
plt.legend()
```

```python
#ndv
def convolution(x,h):
        len_x=np.size(x)
        len_h=np.size(h)
        C = np.zeros(len_x + len_h -1)
        for m in np.arange(len_x):
            for n in np.arange(len_h):
                C[m+n] = C[m+n] + x[m]*h[n]
        return C 
```
