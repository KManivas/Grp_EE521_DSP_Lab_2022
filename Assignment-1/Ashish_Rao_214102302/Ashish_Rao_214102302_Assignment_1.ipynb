```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def step(a,t1,t):
    i=[]
    for temp in t:
        
        if(temp<t1):
            i.append(0)
        else:
            i.append(a)
    return i
```

```python
def ramp(a,t1,t):
    i=[]
    for temp in t:
        if(temp<t1):
            i.append(0)
        else:
            i.append(a*temp)
    return i
```

```python
def parabolic(a,t1,t):
    i=[]
    for temp in t:
        if(temp<t1):
            i.append(0)
        else:
            i.append((a*(temp**2))/2)
    return i
    
```

```python
def pulse(a,t1,t,T):
    i =[]
    for temp in t:
        if temp < t1-T/2:
            i.append(0)
        elif temp > t1+T/2:
            i.append(0)
        else:
            i.append(a)
    return i
```

```python
def convolution(l1,l2):
    m = len(l1)
    n = len(l2)
    result = []
    for i in range(1,m+1):
        result.append(sum(np.multiply(l1[0:i],l2[n-i:n])))
    for j in range(1,n):
        result.append(sum(np.multiply(l1[j:m],l2[0:n-j])))
    return result
```

```python
def sinc(t):
    i = []
    for temp in t:
        y = np.sin(np.pi*temp)/(np.pi*temp)
        i.append(y)
    return i
```

```python
def x_7(t,d1,d2,d3):
    result = []
    for temp in t:
        if temp < 0:
            result.append(0)
        elif temp >= d1 and temp <= d2:
            result.append((temp-d1)/(d2-d1))
        elif temp > d2 and temp <= d3:
            result.append((d3-temp)/(d3-d2))
        elif temp > d3:
            result.append(0)
        else:
            result.append(0)
    return result
```

```python
def real_value(alpha,omegha,phi,t):
    i = []
    for temp in t:
        y = alpha*np.cos((omegha*temp)+phi)
        i.append(y)
    return i

def imag_value(alpha,omegha,phi,t):
    i = []
    for temp in t:
        y = alpha*np.sin((omegha*temp)+phi)
        i.append(y)
    return i

def mag(alpha,omegha,phi,t):
    i = [];
    for temp in t:
        y = alpha*(np.exp(1j*((omegha*temp)+phi)))
        i.append(y);
    return np.abs(i)

def angle(alpha,omegha,phi,t):
    i= [];
    for temp in t:
        y = alpha*(np.exp(1j*((omegha*temp)+phi)))
        i.append(y);
    return np.angle(i)
```

```python
def x_7(t,d1,d2,d3):
    result = []
    for temp in t:
        if temp < 0:
            result.append(0)
        elif temp >= d1 and temp <= d2:
            result.append((temp-d1)/(d2-d1))
        elif temp > d2 and temp <= d3:
            result.append((d3-temp)/(d3-d2))
        elif temp > d3:
            result.append(0)
        else:
            result.append(0)
    return result
```

```python
def convolution(x1,x2):
    result = []
    x3 = []
    for i in range(len(x2)):
        x3.append(x1*x2[i])
    for i in range(len(x3)):
        for j in range(len(x3)-1-i):
            x3[i] = np.insert(x3[i],len(x3[i]),0)
    for i in range(len(x3)):
        for j in range(i):
            x3[i] = np.insert(x3[i],0,0)
    for i in range(len(x3[0])):
        add = 0
        for j in range(len(x3)):
            add+=x3[j][i]
        result.append(add)
    return result


```

```python
# function for real part of complex exponential sequence.
def comp_exp_real(a,b,n):
    com_exp_r =[]
    for sample in n:
        y = np.exp(a*sample)*np.cos(b*sample)
        com_exp_r.append(y)
    return com_exp_r



# function for imaginary part of complex exponential sequence.
def comp_exp_imaginary(a,b,n):
    com_exp_r =[]
    for sample in n:
        y = np.exp(a*sample)*np.sin(b*sample)
        com_exp_r.append(y)
    return com_exp_r
```

```python
def step(a,c,n):
    unit =[]
    for sample in n:
        if sample < c:
            unit.append(0)
        else:
            unit.append(int(a**(-sample+c)))
    return unit
```

```python
def diff_equation(x):
    diff = []
    for i in range(len(x)):
        if i == 0:
            diff.append(x[i])
        else:
            diff.append(x[i]-2*diff[i-1])
    return diff
```
