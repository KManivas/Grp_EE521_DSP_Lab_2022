```python
#Required Imports
import numpy as np
import matplotlib.pyplot as plt
import cmath
```

```python
#function for convolution
def convolution(x,h):
        len_x=np.size(x)
        len_h=np.size(h)
        C = np.zeros(len_x + len_h -1)
        for m in np.arange(len_x):
            for n in np.arange(len_h):
                C[m+n] = C[m+n] + x[m]*h[n]
        return C 
```

#### Unit Step Signal
* $$ x_{1}(t) = u(t) =\begin{cases} 
    0 & \text{$t$ <0}\\ 
    1 & \text{0 $\leq$ $t$ $\leq20$ }\\ 
    \end{cases} $$
  
#### Unit Step Signal with amplitude scaling a=4 and time shift c=-2
* $$ x_{1}(t) = 4u(t-2) $$

#### Unit Step Signal with time scaling b=0.5 and time shift c=-1
* $$ x_{1}(t) = u(0.5t-1) $$

#### Unit Step Signal with time scaling b=-1
* $$ x_{1}(t) = u(-t) $$

```python
#function for unit step signal
def u(t):
    "Unit step signal-->u(t)"
    return 1*(t >= 0)

#unit step signal
t = np.linspace(-3,3,1000)
plt.plot(t,u(t),'blue')
plt.axhline(0,color='black',linewidth = 0.5)
plt.axvline(0,color='black',linewidth = 0.5)
plt.ylim((-0.30,2.20))
plt.title('Unit step signal-->u(t)',color='red',size=20)
plt.tight_layout()
plt.show()

#unit step signal with a=4,c=-2
plt.plot(t,(4*u(t-2)),'blue')
plt.axhline(0,color='black',linewidth = 0.5)
plt.axvline(0,color='black',linewidth = 0.5)
plt.ylim((-0.30,5)) 
plt.title('Unit step signal-->(4*u(t-2))',color='red',size=20)
plt.tight_layout()
plt.show()

#unit step signal with b=0.5,c=-1
plt.plot(t,(u(0.5*t-1)),'blue')
plt.axhline(0,color='black',linewidth = 0.5)
plt.axvline(0,color='black',linewidth = 0.5)
plt.ylim((-0.30,2))
plt.title('Unit step signal-->(u(0.5*t-1))',color='red',size=20)
plt.tight_layout()
plt.show()

#unit step signal with b=-1
plt.plot(t,(u(-t)),'blue')
plt.axhline(0,color='black',linewidth = 0.5)
plt.axvline(0,color='black',linewidth = 0.5)
plt.ylim((-0.30,2))
plt.title('Unit step signal-->(u(-t))',color='red',size=20)
plt.tight_layout()
plt.show()
```

### Unit Ramp Signal
$$ x_{2}(t) = r(t) = tu(t) $$

#### Unit Ramp Signal with amplitude scaling a=4 and time shift c=-2
* $$ x_{2}(t) = 4r(t-2) $$

#### Unit Step Signal with time scaling b=0.5 and time shift c=-1
* $$ x_{2}(t) = r(0.5t-1) $$

#### Unit Step Signal with time scaling b=-1
* $$ x_{2}(t) = r(-t) $$

```python
#ramp signal
def r(t):
    "Ramp signal-->r(t)"
    return t*(1*(t >= 0))
t = np.linspace(-3,3,1000)

#Unit Ramp Signal
plt.plot(t,r(t),'blue')
plt.axhline(0,color='black',linewidth = 0.5)
plt.axvline(0,color='black',linewidth = 0.5)
plt.ylim((-0.30,2.20))
plt.title("Ramp signal-->r(t)",color='red',size=20)
plt.tight_layout()
plt.show()

#Unit Ramp Signal with a=4 and c=-2
plt.plot(t,(4*r(t-2)),'blue')
plt.axhline(0,color='black',linewidth = 0.5)
plt.axvline(0,color='black',linewidth = 0.5)
plt.ylim((-0.30,4))
plt.title("Ramp signal-->(4*r(t-2))",color='red',size=20)
plt.tight_layout()
plt.show()

#Unit Ramp Signal with b=0.5 and c=-1
t = np.linspace(-3,5,1000)
plt.plot(t,(r(0.5*t-1)),'blue')
plt.axhline(0,color='black',linewidth = 0.5)
plt.axvline(0,color='black',linewidth = 0.5)
plt.ylim((-0.30,2.20))
plt.title("Ramp signal-->r(0.5*t-1)",color='red',size=20)
plt.tight_layout()
plt.show()

#Unit Ramp Signal with b=-1
t = np.linspace(-3,5,1000)
plt.plot(t,(r(-t)),'blue')
plt.axhline(0,color='black',linewidth = 0.5)
plt.axvline(0,color='black',linewidth = 0.5)
plt.ylim((-0.30,2.20))
plt.title("Ramp signal-->r(-t)",color='red',size=20)
plt.tight_layout()
plt.show()
```

#### Unit Parabolic Signal
$$ x_{3}(t) = p(t) = t^{2}u(t)/2 $$

#### Unit Parabolic Signal with amplitude scaling a=4 and time shift c=-2
* $$ x_{3}(t) = 4p(t-2) $$

#### Unit Parabolic Signal with time scaling b=0.5 and time shift c=-1
* $$ x_{3}(t) = p(0.5t-1) $$
#### Unit Parabolic Signal with time scaling b=-1
* $$ x_{3}(t) = p(-t) $$

```python
#parabolic signal
def p(t):
    "Parabolic signal-->p(t)"
    return ((t**2)/2)*1*(t >= 0)

#Unit Parabolic Signal
t = np.linspace(-3,3,1000)
plt.plot(t,p(t),'blue')
plt.axhline(0,color='black',linewidth = 0.5)
plt.axvline(0,color='black',linewidth = 0.5)
plt.ylim((-0.30,2.20))
plt.title("Parabolic signal-->p(t)",color='red',size=20)
plt.tight_layout()
plt.show()


#Unit Parabolic Signal with amplitude scaling a=4 and time shift c=-4
plt.plot(t,(4*p(t-2)),'blue')
plt.axhline(0,color='black',linewidth = 0.5)
plt.axvline(0,color='black',linewidth = 0.5)
plt.ylim((-0.30,1))
plt.title("Parabolic signal-->4*p(t-2)",color='red',size=20)
plt.tight_layout()
plt.show()


#Unit Parabolic Signal with time scaling b=0.5 and time shift c=-1
t = np.linspace(-3,5,1000)
plt.plot(t,(p(0.5*t-1)),'blue')
plt.axhline(0,color='black',linewidth = 0.5)
plt.axvline(0,color='black',linewidth = 0.5)
plt.ylim((-0.30,1))
plt.title("Parabolic signal-->p(0.5*t-1)",color='red',size=20)
plt.tight_layout()
plt.show()

#Unit Parabolic Signal with time scaling b=-1 
t = np.linspace(-3,5,1000)
plt.plot(t,(p(-t)),'blue')
plt.axhline(0,color='black',linewidth = 0.5)
plt.axvline(0,color='black',linewidth = 0.5)
plt.ylim((-0.30,1))
plt.title("Parabolic signal-->p(-t)",color='red',size=20)
plt.tight_layout()
plt.show()
```

### Rectangular Signal
$$ x_{4}(t) = rect(t/T)  $$

#### Rectangular Signal with amplitude scaling a=4 and time shift c=-2
* $$ x_{4}(t) = 4rect(t/T-2) $$

```python
#Rectangular pulse with T=1
T=1
def rec(t):
    "Rectangular pulse-->rec(t/T)"
    return (t>=(-T/2))^(t>=(T/2))

#Rectangular signal
t = np.linspace(-3,3,1000)
plt.plot(t,rec(t),'blue')
plt.axhline(0,color='black',linewidth = 0.5)
plt.axvline(0,color='black',linewidth = 0.5)
plt.ylim((-0.30,2.20))
plt.title("Rectangular pulse-->rec(t/T)",color='red',size=20)
plt.tight_layout()
plt.show()

#Rectangular Signal with amplitude scaling a=4 and time shift c=-2
plt.plot(t,(4*rec(t-2)),'blue')
plt.axhline(0,color='black',linewidth = 0.5)
plt.axvline(0,color='black',linewidth = 0.5)
plt.ylim((-0.30,4))
plt.title("Rectangular pulse-->#4rec((t/T)-2)",color='red',size=20)
plt.tight_layout()
plt.show()

```

#### Triangular Signal
* $$ x_{7}(t) =\begin{cases} 
    0 & \text{$t$ <0}\\ 
    (t-d1)/(d2-d1) & \text{d1 $\leq$ $t$ $\leq d2$ }\\
    (d3-t)/(d3-d2) & \text{d2 $\leq$ $t$ $\leq d3$ }\\
    0 & \text{$t$ >15}\\ 
    \end{cases} $$

```python
#triangular signal
def tri(d1,d2,d3):
    "Triangular pulse-->tri(t/T)"
    wf=[]
    i=0
    while i<=20:
        if (i>=3 and i<=10):
            res=((i-d1)/(d2-d1))
        elif (i>=10 and i<=15):
            res=((d3-i)/(d3-d2))
        else:
            res=0
        wf.append(res)
        i=i+0.002
    return wf


#plotting triangular signal with (d1,d2,d3)=(3,10,15)
t = np.arange(0,20.002,0.002)
plt.plot(t,tri(3,10,15),'blue')
plt.axhline(0,color='black',linewidth = 0.5)
plt.axvline(0,color='black',linewidth = 0.5)
plt.title( "Triangular pulse-->tri(t/T)",color='red',size=20)
plt.tight_layout()
plt.grid()
plt.show()
```

## Discrete Time Signals
$$ x_{1}[n] = \delta[n] $$
$$ x_{2}[n] = \delta[3n] $$
$$ x_{3}[n] = \delta[3n-2] $$
$$ x_{4}[n] = \delta[n^{2}+3n+2] $$

#### Consider an input x[n] and a unit impulse response h[n] given by
$$ x[n] = (1/3)^{-n} u[−n − 1] $$
$$ h[n] = u[n − 1] $$
#### Determine and plot the output y[n] = x[n] ∗ h[n]

```python
def u(t):
    "Unit step signal-->u(t)"
    return 1*(t >= 0)


def x1(t):
    return ((1/3)**(-t))*(u(-t-1))
def x2(t):
    return u(t-1)

def convo1(x):
  fog=0
  xt=-5
  dx=1
  while xt<5:
    fog=fog+x1(xt)*x2(x-xt)*dx
    xt=xt+dx
  return(fog)


t = np.arange(-5,5,1)
plt.stem(t,x1(t),'red',markerfmt='ro',label='x1')
plt.stem(t,x2(t),'blue',markerfmt='bo',label='x2')
plt.stem(t,convo1(t),'green',markerfmt='go',label='convolution output')
plt.legend()
```

#### Consider the evaluation of y[n] = x1[n] ∗ x2[n] ∗ x3[n], where
$$ x1 = (0.5)^{n}u[n] $$
$$ x2 = u[n + 3] $$
$$ x3 = δ[n] − δ[n − 1] $$
#### (i) Evaluate the convolution x1[n] ∗ x2[n].
#### (ii) Convolve the result of part (i) with x3[n] in order to evaluate y[n].
#### (iii) Evaluate the convolution x2[n] ∗ x3[n].
#### (iv) Convolve the result of part (iii) with x1[n] in order to evaluate y[n].

```python
def u(t):
    "Unit step signal-->u(t)"
    return 1*(t >= 0)

def impulse(x):
    return 1 * (x == 0)

#x1 fn
def x1(t):
    return (0.5**t)*(t>=0)
#x2 fn
def x2(t):
    return u(t+3)

#x3 fn
def x3(t):
    return (impulse(t)-impulse(t-1))
t = np.arange(-4,4,1)
plt.stem(t,x1(t),'red',markerfmt='ro',label='x1')
plt.stem(t,x2(t),'blue',markerfmt='bo',label='x2')
plt.stem(t,x3(t),'green',markerfmt='go',label='x3')
plt.legend()
plt.show()
```
