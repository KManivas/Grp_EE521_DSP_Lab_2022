```python
#importing libraries:
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
fig, ax=plt.subplots(4,3,figsize=(20,15))        # for subplot and its fig size      
t=np.arange(-20,20,1)                            #range of time axis is from -20 to 20 with step size of 1
x=np.zeros(len(t))                               #taking zeros in variable x  

#UNIT STEP SIGNAL(X1):
for i in range(0,len(t)):
    if t[i]<=0:                                  # conditions for unit step signal as per given limits
        x[i]=0
    elif t[i]>0 and t[i]<=20:
        x[i]=1
# plotting subplot for x1 signal        
ax[0][0].step(t,x,'r')        
ax[0][0].set_title("X1 signal")
ax[0][0].grid()

#RAMP SIGNAL (X2): 
for i in range(0,len(t)):
    if t[i]<=0:
        x[i]=0
    elif t[i]>0 :
        x[i]=t[i]
# plotting subplot for x2 signal        

ax[0][1].plot(t,x,'y')
ax[0][1].set_title("X2 signal")
ax[0][1].grid()

# PARABOLIC SIGNAL(X3):

for i in range(0,len(t)):
    if t[i]<=0:
        x[i]=0
    elif t[i]>0 :
        x[i]=(t[i]**2)/2
# plotting subplot for x3 signal        
ax[0][2].plot(t,x,color='purple')
ax[0][2].set_title("X3 signal")
ax[0][2].grid()

#RECTANGULAR SINAL(X4):
tau=20                                                   # variable for rect(t/tau)
for i in range(0,len(t)):
    if np.abs(t[i])<= (-tau/2) or np.abs(t[i])>= (tau/2):
        x[i]=0
    else:
        x[i]=1
        
# plotting subplot for x4 signal        
        
ax[1][0].step(t,x,color='orange')
ax[1][0].set_title("X4 signal")
ax[1][0].grid()   

#convolution OF RECTANGULAR SIGNAL(X5):  

def con(u,v):                                                       #CONVOLUTION FUNCTION DEFINED HERE
    size=len(u)+len(v)-1
    newv=np.pad(v,(0,size-len(v)))
    sum=np.zeros(size)
    for i in range(len(u)):
        sum=sum+u[i]*newv
        newv=np.roll(newv,1)
    return sum  

x5= con(x,x)
ax[1][1].plot(np.arange(-40,39,1),x5,color='maroon')
ax[1][1].set_title("X5 signal")
ax[1][1].grid()   

#SINC SIGNAL(X6):
t1=np.linspace(-10,10,1000)
x6= np.sin(3.14*t1)/(3.14*t1)
ax[1][2].plot(t1,x6,color='maroon')
ax[1][2].set_title("X6 signal")
ax[1][2].grid()   

# X7 SIGNAL:

d1,d2,d3=3,10,15
for i in range(0,len(t)):
    if t[i]<d1:                                   # applying conditions for x7 sig
        x[i]=0
    elif t[i]>=d1 and t[i]<=d2:
        x[i]=(t[i]-d1)/(d2-d1)
    elif t[i]>=d2 and t[i]<=d3:
        x[i]=(d3-t[i])/(d3-d2) 
    elif  t[i]>=d3 :
        x[i]=0
ax[2][0].plot(t,x,color='green')
ax[2][0].set_title("X7 signal")
ax[2][0].grid()   


#             ----TIME SHIFTED AND AMPLITUDE SCALED VERSION OF X1,X2,X3,X4 & X5 SIGNALS----

c=int(input("TIME SHIFT="))
a=int(input("AMPLITUDE SCALING FACTOR="))
# signal x1:
for i in range(0,len(t)):
    if t[i]<=c:
        x[i]=0
    elif t[i]>c and t[i]<=(20+c):
        x[i]=a
ax[2][1].step(t,x,'r')
ax[2][1].set_title("Amp scaled & time shifted X1 signal")
ax[2][1].grid() 

#signal x2:

for i in range(0,len(t)):
    if t[i]<=c:
        x[i]=0
    elif t[i]>c :
        x[i]=a*(t[i]-c)

ax[2][2].plot(t,x,'y')
ax[2][2].set_title("Amp scaled & time shifted X2 signal")
ax[2][2].grid()
#signal x3:
for i in range(0,len(t)):
    if t[i]<=c:
        x[i]=0
    elif t[i]>c :
        x[i]= a*((t[i]-c)**2)/2
ax[3][0].plot(t,x,'y')
ax[3][0].set_title("Amp scaled & time shifted X3 signal")
ax[3][0].grid()

#signal x4:
for i in range(0,len(t)):
    if t[i]>=(-tau/2)+c and np.abs(t[i])<=(tau/2)+c:
        x[i]=a
    else:
        x[i]=0
ax[3][1].step(t,x,'y')
ax[3][1].set_title("Amp scaled & time shifted X4 signal")
ax[3][1].grid()

#signal x5:
x5=con(x,x)
ax[3][2].plot(np.arange(-40,39,1),x5,color='purple')             #since after convolution range of sig increases, that's why x-axis range is defined in this instruction
ax[3][2].set_title("Amp scaled & time shifted X5 signal")
ax[3][2].grid()
plt.show()
```

```python
n=np.arange(-2,5,1)
x=np.zeros(len(n))
# function for impulse signal with a:amplitude, b:time scale, c:time shift ;
def h(a,b,c):
    for i in range(0,len(n)):
            if (n[i])==(c/b):
                x[i]=a
            else:
                x[i]=0
               
    return np.array(x)
# generating sgnals from above function:

x1=h(3,1,0)+h(5,1,1)+h(2,1,2)+h(6,1,3)+h(8,1,4)
x2=h(2,1,0)+h(3,1,1)+h(4,1,2)+h(7,1,3)+h(1,1,4)
h1=h(-2,1,0)+h(4,1,1)+h(1,1,2)
h2=h(-2,-1,0)+h(4,-1,1)+h(1,-1,2)
# performing convolution using numpy

c1=con(x1,h1)
c2=con(x2,h2)
print("(i) X[n]:",x1," and h[n]:",h1)
print("x[n]*h[n]:",c1)
print("(ii)X[n]:",x2," and h[n]:",h2)
print("x[n]*h[n]:",c2)

#plotting convolved signal:
fig, ax=plt.subplots(1,2,figsize=(11,7))        
ax[0].stem(n,x1,'y',label='x[n]:yellow')
ax[0].stem(n,h1,'b',label='h[n]:blue')
ax[0].stem(np.arange(-2,11,1),c1,'r',label='x[n]*h[n]:red')
ax[0].set_title("(i)x[n]*h[n]")
ax[0].grid()
ax[0].legend()
ax[1].stem(n,x2,'r',label='x[n]:red')
ax[1].stem(n,h2,'b',label='h[n]:blue')
ax[1].stem(np.arange(-2,11,1),c2,'y',label='x[n]*h[-n]:yellow')
ax[1].set_title("(i)x[n]*h[-n]")
ax[1].legend()
ax[1].grid()
```

```python
#For impulse response x[n]= impulse[n] and y[n]=h[n]:
n=np.arange(-1,7,1)
x=np.zeros(len(n))
IR=h(3,1,1)    #impulse response of system
fig, ax=plt.subplots(2,4,figsize=(20,15)) 

#x[n]=u[n] and h[n]=IR ;y[n]=x[n]*h[n]

u=unit(1,1,0)
ax[0][0].stem(n,u,'r')
ax[0][0].set_title("input:unit step")
ax[0][0].grid()
#UR:unit resonse
UR=con(IR,u)
ax[0][1].stem(np.arange(-5,10,1),UR,'r')
ax[0][1].set_title("output for unit step")
ax[0][1].grid()

#x[n]=r[n] and h[n]=IR ;y[n]=x[n]*h[n]

r=ramp(1,1,0)
ax[0][2].stem(n,r,'blue')
ax[0][2].set_title("input:ramp signal")
ax[0][2].grid()
# RR: ramp response
RR=con(IR,r)
ax[0][3].stem(np.arange(-5,10,1),RR,'blue')
ax[0][3].set_title("output for ramp")
ax[0][3].grid()

#x[n]=p[n] and h[n]=IR ;y[n]=x[n]*h[n]

p=par(1,1,0)
ax[1][0].stem(n,p,'g')
ax[1][0].set_title("input:parabolic signal")
ax[1][0].grid()
#PR: parabolic response 
PR=con(IR,p)
ax[1][1].stem(np.arange(-5,10,1),PR,'g')
ax[1][1].set_title("output for parabolic signal")
ax[1][1].grid()

#x[n]=real part of exp signal and h[n]=IR ;y[n]=x[n]*h[n]

real_in=np.exp(2*n)*np.cos(3*n)
ax[1][2].stem(n,real_in,'g')
ax[1][2].set_title("input:complex exp real part")
ax[1][2].grid()
real_out=con(real_in,IR)
ax[1][3].stem(np.arange(-5,10,1),real_out,'g')
ax[1][3].set_title("output for real part")
ax[1][3].grid()
```

IN real part of complex discrete exp signal:

output response is advanced by 2 samples w.r.t input and there is increment in amplitude(pattern of the sequence is same as input) .

IN imaginary part of complex discrete exp signal:

```python
n=np.arange(-5,20,1)
x=np.zeros(len(n))
fig, ax=plt.subplots(1,3,figsize=(15,5)) 

# function for given signal:

def X(b,c):
    for i in range(len(n)):
        if(n[i]/b)<c:
            x[i]=0
        elif (n[i]/b)>=c :
            x[i]=(1/3)**(-n[i])
    return np.array(x)

u=np.array(unit(1,1,1))
y=con(u,X(-1,1))
print("OUTPUT y[n]:",y)
ax[0].stem(n,X(-1,1),'r')
ax[0].set_title("x[n]")
ax[0].grid()
ax[1].stem(n,unit(1,1,1),'g')
ax[1].set_title("h[n]")
ax[1].grid()
ax[2].stem(np.arange(-10,39,1),y,'purple')
ax[2].set_title("y[n]=x[n]*h[n]")
ax[2].grid()
plt.show()
```

Y(Z)(1+(2/z))=X(Z)   -------------(1)

in z-transform H(Z)=Y(Z)/X(Z)

H(Z)=z/(z+2)        --------------(from eq 1)

By inverse z-transform;

h[n]= ((-2)^n)u[n]




```python
n=np.arange(-2,10,1)
x=np.zeros(len(n))
# calculation for given signal
for i in range(len(n)):
            if(n[i])<0:
                x[i]=0
            elif (n[i])>=0 :
                x[i]=(-2)**(n[i])
                
plt.stem(n,x,'g')
plt.title("Impulse response for given difference eq")
plt.grid()
```
