# Continuous Time signals

```python
t=np.arange(-10,50,0.1)
fig,axes=plt.subplots(nrows=1,ncols=3,figsize=(14,6))
def x1(t):
    return 1 *(t>=0)*(t<=20)



axes[0].set_title('x1(t) ')
axes[0].set_xlabel('t')
axes[0].set_ylabel('u(t)')
axes[0].set_xlim(-10,40)
axes[0].plot(t,x1(t))



axes[1].set_title('4*x1(t-2)')
axes[1].set_xlabel('t')
axes[1].set_ylabel('4*u(t-2)')
axes[1].set_xlim(-10,40)
axes[1].plot(t,4*x1(t-2))

axes[2].set_title('x1(0.5*t-1)')
axes[2].set_xlabel('t')
axes[2].set_ylabel('u(0.5*t-1)')
axes[2].plot(t,x1(0.5*t-1))
plt.tight_layout()
```

```python
t=np.arange(-10,50,1)
fig,axes=plt.subplots(nrows=1,ncols=3,figsize=(14,6))
def x2(t):
    return t *(t>=0)*(t<=20)


axes[0].set_xlabel('t')
axes[0].set_ylabel('x2(t)')
axes[0].set_title('x2(t)')
axes[0].set_xlim(-10,40)
axes[0].plot(t,x2(t))

axes[1].set_title('4*x2(t-2)')
axes[1].set_xlabel('t')
axes[1].set_ylabel('4*r(t-2)')
axes[1].set_xlim(-10,40)
axes[1].plot(t,4*x2(t-2))


axes[2].set_title('x2(0.5*t-1)')
axes[2].set_xlabel('t')
axes[2].set_ylabel('r(0.5*t-1)')
axes[2].plot(t,x2(0.5*t-1))

plt.tight_layout()
```

```python
t=np.arange(-10,50,1)
fig,axes=plt.subplots(nrows=1,ncols=3,figsize=(14,6))
def x3(t):
    return (np.square(t)/2) * (t >= 0)*(t<=20)

axes[0].set_xlabel('t')
axes[0].set_ylabel('x3(t)')
axes[0].set_title('p(t)')
axes[0].plot(t,x3(t))
axes[0].grid()

axes[1].set_title('4*x3(t-2)')
axes[1].set_xlabel('t')
axes[1].set_ylabel('4*p(t-2)')
axes[1].plot(t,4*x3(t-2))
axes[1].grid()


axes[2].set_title(r'x3(0.5*t-1)')
axes[2].set_xlabel('t')
axes[2].set_ylabel('p(0.5*t-1)')
axes[2].plot(t,x3(0.5*t-1))
axes[2].grid()

plt.tight_layout()
```

```python
t=np.linspace(-10,30,500)
d_1=3
d_2=10
d_3=15
s=[]
for i in range(0,500):
    if (t[i]<0):
        s.append(0)
    elif (t[i]>=d_1) & (t[i]<=d_2):
        s.append((t[i]-d_1)/(d_2-d_1))
    elif (t[i]>=d_2) & (t[i]<=d_3):
        s.append((d_3-t[i])/(d_3-d_2))
    else :
        s.append(0)
plt.figure(figsize=(10,6))
plt.xlabel('t')
plt.ylabel('x7(t)')
plt.title('x7(t)')
plt.plot(t,s)
```

```python
def triangular_wave(a,n,ul,ll,step_size,period):
    tria_wave= []
    Range = ul - ll
    T = Range/period
    for i in range(period):
        t1 = np.arange(ll,ll+T, step_size)
        for j in t1:
            if abs(j) <= T/4:
                tria_wave.append(a * j)
            elif abs(j) > T/4 and abs(j) <= T/2:
                tria_wave.append((a*T/2)-(a*j))
            elif abs(j) > T/2 and abs(j) <= 3*T/4:
                tria_wave.append(((a*T/2)-(a*j)))
            else:
                tria_wave.append(-((a*T)-(a*j)))
    return tria_wave

ll=0
ul=40
step_size=1
period=4
a=2
n=np.arange(ll,ul,step_size)
plt.figure(figsize=(12,6))
plt.stem(n,triangular_wave(a,n,ul,ll,step_size,period))
plt.title('Discrete Triangular wave')
plt.grid()
```

Let x[n] be the input to an linear-time invariant (LTI) system and
h[n] be the impulse response. The output of a discrete LTI system
in terms of the impulse response is characterized by the equation
as
y[n] = x[n] ∗ h[n] = X∞
k=−∞
x[k]h[n − k]
Write a function for performing convolution of two sequences.
By using this function, calculate the output for the following sequences:
(i) x[n] = {3, 5, 2, 6, 8} and h[n] = {−2, 4, 1}
(ii) x[n] = {2, 3, 4, 7, 1} and h[−n] = {−2, 4, 1}
For (i) and (ii), plot x[n], h[n], and y[n] on a single subplot

```python
def convolution(x1,x2):
    result = []
    x3 = []
    for i in range(len(x2)):
        x3.append(x1*x2[i])
    for i in range(len(x3)):
        for j in range(len(x3)-1-i):
            x3[i] = np.insert(x3[i],len(x3[i]),0)
    for i in range(len(x3)):
        for j in range(i):
            x3[i] = np.insert(x3[i],0,0)
    for i in range(len(x3[0])):
        add = 0
        for j in range(len(x3)):
            add+=x3[j][i]
        result.append(add)
    return result
```

```python
x=np.array([3,5,2,6,8])
h=np.array([-2,4,1])
y=convolution(x,h)
fig,axes=plt.subplots(nrows=1,ncols=3,figsize=(14,6))

n1=np.linspace(0,4,len(x))
axes[0].set_title('x[n]')
axes[0].stem(n1,x)

n2=np.linspace(0,2,len(h))
axes[1].set_title('h[n]')
axes[1].stem(n2,h)

n=np.linspace(0,6,len(y))
axes[2].set_title('y[n]')
axes[2].stem(n,y)
```

```python
def step_seq(n,a,c):
    unit_fn = []
    
    for i in n:
        if i > c:
            unit_fn.append(0)
        else:
            unit_fn.append(int(a**(-c+i)))
    return unit_fn
n=np.arange(-10,10,1)
n_2=np.arange(-19,20,1)
u_1=unit_step(1,n)
s_1=step_seq(n,(1/3),-1)
c_1=convolution(u_1,s_1)
plt.figure(figsize=(14,6))
plt.title('y[n]')
plt.stem(n_2,c_1)
```

Consider the first-order difference equation: y[n]+2y[n−1] = x[n].Assuming the condition of initial rest 
(i.e., if x[n] = 0 for n < n0, then y[n] = 0 for n < n0), find the impulse response of a system whose input and 
output are related by this difference equation.


```python
def difference_equation(x):
    diff= []
    for i in range(len(x)):
        if i == 0:
            diff.append(0)
        else:
            diff.append(x[i]-2*diff[i-1])
    return diff
```

```python
n = np.arange(-5,5,1)
x = unit_impulse(0,n)
y = difference_equation(x)

plt.figure(figsize=(14,6))
plt.subplot(1,2,1)
plt.stem(n,x)
plt.title("unit impulse signal")

plt.subplot(1,2,2)
plt.stem(n,y)
plt.title("difference equation output")

plt.tight_layout()
```
