## __$x_5(t) = rect\big(\frac{t}{\tau}\big) * rect\big(\frac{t}{\tau}\big)$__  $ = x_4(t) * x_4(t)$, $4x_5(t - 2)$  , $x_5(0.5t)$ , $x_5(-t)$

```python
'''Convolution operation  of finite length signals:
         
             
            Tabulation method:

'''

def convolve(x1 , x2):       
    
    n1 = len(x1)
    n2 = len(x2)
           
    
    res = []          
    
    for i in range(n1 + n2 - 1):
        
        s = 0
        
        x = 0
        y = i
        
        while x <= i and y >= 0:
            
            if (x >= 0 and  x<n1) and (y>=0 and y<n2):
                s += x1[x]*x2[y]

            x += 1
            y -= 1
                
                  
        res.append(s)

    return res
    
```

```python
x5 = convolve(x4,x4)

t = np.linspace(-6 ,6,len(x5))



#plotting

plt.figure(figsize = (15,7))

plt.subplot(221)
plt.plot(t,x5)
plt.grid()
plt.xlabel(f'$t$' , size=15,weight='bold')
plt.ylabel(f'$x_5(t)$' , size=15,weight='bold')
plt.title(r'$x_5(t)$ = $ rect (\frac{t}{\tau}) * rect (\frac{t}{\tau})$' , size=20,weight = 'extra bold')


plt.subplot(222)
x5_2 =  convolve(x4_2 , x4)
t5_2 = np.linspace(-6 ,6,len(x5_2))
plt.plot(t5_2,x5_2)
plt.grid()
plt.xlabel(f'$t$' , size=15,weight='bold')

plt.title(f'$4x_5(t-2) $' , size=20,weight = 'extra bold')
plt.axvline(0 , ls = '--' , color = 'g' , label = 't = 0')
plt.axvline(2 , ls = '--' , color = 'r' , label = 't = 2')
plt.legend(prop = {'size' : 15})


plt.subplot(223)
plt.plot(t/2,x5)
plt.grid()
plt.xlabel(f'$t$' , size=15,weight='bold')

plt.title(f'$x_5(0.5t) $' , size=20,weight = 'extra bold')

plt.subplot(224)
x5_4 = x5[::-1]
plt.plot(t,x5_4)
plt.grid()
plt.xlabel(f'$t$' , size=15,weight='bold')

plt.title(f'$x_5(-t) $' , size=20,weight = 'extra bold')

plt.tight_layout()
plt.show()

```

### $(\alpha , \omega , \phi) = (1 , 0.14\pi , -\frac{\pi}{3})$ ,$(2 , 0.34\pi , -\frac{\pi}{2}) , (2 , 0.75 \pi , \pi)$

```python
params = [[1 ,0.14 * np.pi ,-1*np.pi/3] , [2 ,0.34*np.pi , -1*np.pi/2] , [2,0.75*np.pi , np.pi]]

t = np.linspace(-10,10,1000)



fig , ax = plt.subplots(4,3,sharex = True , sharey = True)


fig.set_size_inches(15,10)
fig.set_tight_layout('tight')




ax[3][0].set_xlabel('t' , size = 15 , weight = 'bold')
ax[3][1].set_xlabel('t' , size = 15 , weight = 'bold')
ax[3][2].set_xlabel('t' , size = 15 , weight = 'bold')

cl = ['r','g','b']

for j in range(3):
    
    alp , omg , phi = params[j]
    
    real = alp * np.cos(omg*t+phi)
    imag = alp * np.sin(omg*t+phi)
    mag = np.sqrt(real**2 + imag**2)
    ang = np.arctan2(imag , real)
    
    

    ax[0][j].plot(t,real , color = cl[j] )
    ax[0][j].set_title(f'y(t) = {alp}exp(j{round(omg,3)}t + {round(phi,3)}) \n Re(y(t)) ' , size = 17 , weight = 'extra bold')
    ax[0][j].grid()

    ax[1][j].plot(t,imag,color = cl[j] )
    ax[1][j].set_title('Im{y(t)}' , size = 15 , weight = 'extra bold')
    ax[1][j].grid()

    ax[2][j].plot(t,mag,color = cl[j])
    ax[2][j].set_title('|y(t)|' , size = 15 , weight = 'extra bold')
    ax[2][j].grid()


    ax[3][j].plot(t,ang,color = cl[j])
    ax[3][j].set_title(r'$\angle $$y(t)$' , size = 15 , weight = 'extra bold')
    ax[3][j].grid()
        
        
    



```
