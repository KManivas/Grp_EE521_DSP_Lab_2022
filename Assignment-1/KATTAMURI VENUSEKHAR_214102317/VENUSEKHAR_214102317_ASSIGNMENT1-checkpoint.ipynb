# *Basic signals

```python
#x1---unit step signal
def u(t):
    "x1-->unit step signal-->u(t)"
    return 1*(t >= 0)


#x2---ramp signal
def r(t):
    "x2-->ramp signal-->r(t)"
    return t*(1*(t >= 0))


#x3---parabolic signal
def p(t):
    "x3-->parabolic signal-->p(t)"
    return ((t**2)/2)*1*(t >= 0)


#x4----rectangular pulse
def rec(t):
    "x4-->rectangular pulse-->rec(t/T)"
    return (t>=(-T/2))^(t>=(T/2))


#x5----triangular pulse
def tri(t):
    "x5-->triangular pulse-->tri(t/T)"
    return (1 -(abs(t)/T))*(abs(t)<T)



#x6---sinc function
def sinc(t):
    "x6-->sinc function-->sinc(t)"
    return  np.sin(np.pi*t)/(np.pi*t)


T=1;
functions = [u,r,p,rec,tri,sinc]

t = np.linspace(-2,2,1000)
fig = plt.figure(10,figsize=(14,8))

for i,function in enumerate(functions,start=1):
    plt.subplot(3,2,i)
    plt.plot(t,function(t),'blue')
    plt.axhline(0,color='black',linewidth = 0.5)
    plt.axvline(0,color='black',linewidth = 0.5)
    plt.ylim((-0.30,2.20))
    plt.title(function.__doc__,color='red',size=20)
    plt.grid()
    
plt.tight_layout()
plt.show()
```

```python
#x1(n)

def i1(a, n):
    "impulse(n)"
    delta1 =[] 
    for sample in n: 
        if sample == a: 
            delta1.append(1) 
        else: 
            delta1.append(0) 
    return delta1 

#x2(n)

def i2(a, n):
    "4*impulse(3n)"
    delta2 =[] 
    for sample in n: 
        if sample == a: 
            delta2.append(3) 
        else: 
            delta2.append(0) 
    return delta2

#x3(n)

def i3(a, n):
    "4*impulse(3n-2)"
    delta3 =[] 
    for sample in n: 
        if (3*sample-2) == a: 
            delta3.append(4) 
        else: 
            delta3.append(0) 
    return delta3

#x4(n)

def i4(a, n): 
    "4*impulse(n**2+3n+2)"
    delta4 =[] 
    for sample in n: 
        if sample == -2: 
            delta4.append(4)
        elif sample == -3:
            delta4.append(4)
        else: 
            delta4.append(0) 
    return delta4


#x5(n)

def i5(a, n):
    "x5[n] = x1[n] + x2[n] + x3[n]"
    delta5=[]
    for sample in n:
        if sample == 0: 
            delta5.append(5)
        else: 
            delta5.append(0)
    return delta5

a=0;functions = [i1,i2,i3,i4,i5]

n = np.arange(-10,10,1)

fig = plt.figure(10,figsize=(10,14))

for i,f in enumerate(functions,start=1):
    plt.subplot(5,1,i)
    plt.stem(n,f(a,n),use_line_collection=True)
    plt.title(f.__doc__,color='red',size=20)
plt.tight_layout()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def convolution(inputseq,impulseresponse):
        lengthA=np.size(inputseq)
        lengthB=np.size(impulseresponse)
        C = np.zeros(lengthA + lengthB -1)
        for m in np.arange(lengthA):
            for n in np.arange(lengthB):
                C[m+n] = C[m+n] + inputseq[m]*impulseresponse[n]
        return C 
    
```

```python
fig = plt.figure(1,figsize=(10,7))

a=[3, 5, 2, 6, 8]
b=[-2, 4, 1]
n = np.arange(-10,10,1)
p=convolution(a,b)
print("output sequence=",p)

plt.subplot(3,1,1)
plt.stem(p,use_line_collection=True)
plt.title('output',color='red',size=20)

plt.subplot(3,1,2)
plt.stem(a,use_line_collection=True)
plt.title('input',color='red',size=20)

plt.subplot(3,1,3)
plt.stem(b,use_line_collection=True)
plt.title('impulse sequence',color='red',size=20)

plt.tight_layout()
```

```python
fig = plt.figure(1,figsize=(10,7))

a=[2, 3, 4, 7, 1]
b=[-2, 4, 1]
n = np.arange(-10,10,1)
p=convolution(a,b)
print("output sequence=",p)

plt.subplot(3,1,1)
plt.stem(p,use_line_collection=False)
plt.title('output',color='red',size=20)

plt.subplot(3,1,2)
plt.stem(a,use_line_collection=True)
plt.title('input',color='red',size=20)

plt.subplot(3,1,3)
plt.stem(b,use_line_collection=True)
plt.title('impulse sequence',color='red',size=20)

plt.tight_layout()
```
