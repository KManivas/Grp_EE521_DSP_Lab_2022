```python
x = np.linspace(-np.pi ,np.pi,1000 )

f = (np.pi - x) * (x>=0)


plt.figure(figsize = (12,8))
plt.plot(x,f)
plt.title('f(x)' , size= 20,weight = 'extra bold')
plt.xlabel('x' , size = 15 , weight = 'bold')
plt.grid()


T = 2 * np.pi

a0 = (1/T) * quad(lambda t:np.pi - t , 0 , np.pi)[0]
plt.axhline(a0 , ls = '--' , c = 'r' , label = f'$a0$ = {a0}')
plt.xlim(-np.pi , np.pi)
plt.legend(prop = {'size':15})
plt.show()

an = []
bn = []

for i in range(1,31):
    
    ai = (2/T) * quad(lambda t:(np.pi - t)*np.cos(i*t) , 0 , np.pi)[0]
    bi = (2/T) * quad(lambda t:(np.pi - t)*np.sin(i*t) , 0 , np.pi)[0]
    
    an.append(ai)
    bn.append(bi)

print('Fourier series coefficients:\n')
for i in range(1,31):
    print(f'\ta{i} = {an[i-1]}\t\tb{i} = {bn[i-1]}')
 

plt.figure(figsize = (12,8))
plt.stem(np.arange(1,31) , an ,linefmt = 'g', markerfmt = 'ob' , label = r'$ an = \frac{2}{\pi n^2}(1 - (-1)^n) $' ,)
plt.stem(np.arange(1,31) , bn , markerfmt = '^r',label = r'$bn = \frac{1}{n}$'  )
plt.xticks(np.arange(1,31))
plt.grid()
plt.title('Plot of Fourier series coefficients   $an ,bn$' , size= 15,weight = 'extra bold')
plt.legend(prop = {'size':18})
plt.show()
```

```python
x_2 = [-1,0,1]
f_2 = [-1,1,-1]

plt.figure(figsize = (12,8))
plt.plot(x_2,f_2)
plt.title('f(x)' , size= 18,weight = 'extra bold')
plt.xlabel('x' , size = 15 , weight = 'bold')
plt.grid()

T = 2
omg = (2 * np.pi)/T

Q2_a0 = (1/T) * (quad(lambda t:2*t + 1 , -1 , 0)[0] + quad(lambda t:-2*t + 1 , 0 , 1)[0])

plt.axhline(Q2_a0 , ls = '--' , c = 'r' , label = f'$a_0$ = {round(Q2_a0,3)}')
plt.xlim(-1 , 1)
plt.legend(prop = {'size':15})
plt.show()

Q2_an = []
Q2_bn = []

for i in range(1,16):
    
    ai = (2/T) * (quad(lambda t:(2*t + 1) * np.cos(i*omg*t) , -1 , 0)[0] + quad(lambda t:(-2*t + 1) * np.cos(i*omg*t) , 0 , 1)[0])
    bi = (2/T) * (quad(lambda t:(2*t + 1) * np.sin(i*omg*t) , -1 , 0)[0] + quad(lambda t:(-2*t + 1) * np.sin(i*omg*t) , 0 , 1)[0])
    
    Q2_an.append(ai)
    Q2_bn.append(bi)
    
print('Fourier series coefficients:\n')
for i in range(1,16):
    print(f'\ta{i} = {Q2_an[i-1]}\t\tb{i} = {Q2_bn[i-1]}')
    

plt.figure(figsize = (15,7))
plt.stem(np.arange(1,16) , Q2_an ,linefmt = 'g', markerfmt = 'ob' , label = r'$ a_n = \frac{4}{\pi^2n^2}(1-(-1)^n) $' ,)
plt.stem(np.arange(1,16) , Q2_bn , markerfmt = '^r',label = r'$b_n = 0 $'  )
plt.xticks(np.arange(1,16))
plt.grid()
plt.title('Plot of Fourier series coefficients   $a_n ,b_n$' , size= 15,weight = 'extra bold')
plt.legend(prop = {'size':18})
plt.show()
```

```python
import cmath

'''Fourier Transform'''
def FT(func , omg , lmts = [-inf,inf]):
    
    l,h = lmts

    Q3_re = []
    Q3_im = []

    for i in omg:
        
        Q3_re.append(quad(eval('lambda t:' + func +f'* np.real(cmath.exp(-1j *{i}*t))') , l,h)[0])
        Q3_im.append(quad(eval('lambda t:' + func +f'* np.imag(cmath.exp(-1j *{i}*t))') , l,h)[0])
        
    
        
    Q3_mag = np.sqrt(np.power(Q3_re,2)+np.power(Q3_im,2))
    Q3_ph = np.arctan2(Q3_im,Q3_re)
    
    return Q3_mag , Q3_ph
```

```python
'''DFT'''
def DFT(x):
    N = len(x)
    
    D = np.zeros((N,N),dtype=np.complex64)
    
    for k in range(N):
        for n in range(N):
            D[k][n] = np.exp(-2j*math.pi*k*n/N)
    
    dft = np.dot(D,x)
    return dft


sum_sin_dft = DFT(sum_sin)

sum_sin_dft_mag = np.apply_along_axis(lambda x:np.absolute(x) , 0 , sum_sin_dft)


k = np.arange(len(sum_sin))
plt.stem(k ,sum_sin_dft_mag , cl[1])
plt.title(r'$\|X(K)|$' , size= 20,weight = 'extra bold')
plt.xlabel('$K$' , size = 15 , weight = 'bold')
plt.grid()
plt.xticks([2,5,8,92,98,95] )
plt.show()
```
